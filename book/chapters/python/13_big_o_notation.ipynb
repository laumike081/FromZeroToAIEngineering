# Big-O Notation

Big-O describes how algorithm performance scales with input size.

---

## Common Complexities

| Big-O | Name | Example |
|-------|------|--------|
| O(1) | Constant | Array access, hash lookup |
| O(log n) | Logarithmic | Binary search |
| O(n) | Linear | Linear search, single loop |
| O(n log n) | Linearithmic | Merge sort, quick sort |
| O(n²) | Quadratic | Nested loops, bubble sort |
| O(n³) | Cubic | Triple nested loops |
| O(2ⁿ) | Exponential | Recursive fibonacci, subsets |
| O(n!) | Factorial | Permutations |

```python
import matplotlib.pyplot as plt
import numpy as np

n = np.arange(1, 20)

plt.figure(figsize=(10, 6))
plt.plot(n, np.ones_like(n), label='O(1)')
plt.plot(n, np.log2(n), label='O(log n)')
plt.plot(n, n, label='O(n)')
plt.plot(n, n * np.log2(n), label='O(n log n)')
plt.plot(n, n**2, label='O(n²)')
plt.plot(n, 2**n, label='O(2ⁿ)')

plt.xlabel('Input Size (n)')
plt.ylabel('Operations')
plt.title('Big-O Complexity Comparison')
plt.legend()
plt.ylim(0, 100)
plt.grid(True)
plt.show()
```

## Time Complexity Examples

```python
# O(1) - Constant
def get_first(arr):
    return arr[0] if arr else None

# O(log n) - Logarithmic
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

# O(n) - Linear
def find_max(arr):
    max_val = arr[0]
    for x in arr:  # n iterations
        if x > max_val:
            max_val = x
    return max_val

# O(n²) - Quadratic
def has_duplicate_pair(arr):
    for i in range(len(arr)):       # n iterations
        for j in range(i+1, len(arr)):  # n iterations
            if arr[i] == arr[j]:
                return True
    return False
```

## Rules for Calculating Big-O

### 1. Drop Constants

```python
def example(arr):
    # 3 operations - still O(1)
    x = arr[0]
    y = arr[1]
    z = arr[2]
    
    # 2n operations - still O(n)
    for i in arr:
        print(i)
    for i in arr:
        print(i)
```

### 2. Drop Non-Dominant Terms

```python
def example(arr):
    # O(n) + O(n²) = O(n²)
    for x in arr:          # O(n)
        print(x)
    
    for x in arr:          # O(n²)
        for y in arr:
            print(x, y)
```

### 3. Different Inputs = Different Variables

```python
def example(arr1, arr2):
    # O(a + b), NOT O(n)
    for x in arr1:  # O(a)
        print(x)
    for y in arr2:  # O(b)
        print(y)

def example2(arr1, arr2):
    # O(a * b), NOT O(n²)
    for x in arr1:
        for y in arr2:
            print(x, y)
```

## Space Complexity

Memory used by the algorithm (excluding input).

```python
# O(1) space - constant extra memory
def sum_array(arr):
    total = 0  # Only one variable
    for x in arr:
        total += x
    return total

# O(n) space - linear extra memory
def double_array(arr):
    result = []  # New array of size n
    for x in arr:
        result.append(x * 2)
    return result

# O(n) space - recursion stack
def factorial(n):
    if n <= 1:
        return 1
    return n * factorial(n - 1)  # n stack frames
```

## Amortized Analysis

Average time per operation over a sequence of operations.

```python
# list.append() is O(1) amortized
# Occasionally O(n) when resizing, but rare

arr = []
for i in range(1000000):
    arr.append(i)  # O(1) amortized

print(f"Length: {len(arr)}")
```

## Key Takeaways

| Concept | Key Point |
|---------|----------|
| **Big-O** | Worst-case upper bound |
| **Drop constants** | O(2n) = O(n) |
| **Drop non-dominant** | O(n² + n) = O(n²) |
| **Space matters** | Consider memory usage |
| **Trade-offs** | Often time vs space |
