# Object-Oriented Programming

OOP organizes code into classes and objects for better structure and reusability.

---

## Classes and Objects

```python
class Dog:
    """A simple Dog class."""
    
    # Class attribute (shared by all instances)
    species = "Canis familiaris"
    
    def __init__(self, name, age):
        """Initialize instance attributes."""
        self.name = name  # Instance attribute
        self.age = age
    
    def bark(self):
        """Instance method."""
        return f"{self.name} says woof!"
    
    def __str__(self):
        """String representation."""
        return f"Dog({self.name}, {self.age})"

# Create instances
buddy = Dog("Buddy", 3)
max_dog = Dog("Max", 5)

print(buddy.bark())
print(f"Species: {buddy.species}")
print(buddy)
```

## Inheritance

```python
class Animal:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        raise NotImplementedError("Subclass must implement")

class Cat(Animal):
    def speak(self):
        return f"{self.name} says meow!"

class Dog(Animal):
    def speak(self):
        return f"{self.name} says woof!"

# Polymorphism
animals = [Cat("Whiskers"), Dog("Rex")]
for animal in animals:
    print(animal.speak())
```

## Special Methods (Dunder Methods)

| Method | Purpose |
|--------|--------|
| `__init__` | Constructor |
| `__str__` | String representation |
| `__repr__` | Developer representation |
| `__len__` | Length (`len(obj)`) |
| `__eq__` | Equality (`==`) |
| `__lt__` | Less than (`<`) |

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)
    
    def __repr__(self):
        return f"Vector({self.x}, {self.y})"
    
    def __eq__(self, other):
        return self.x == other.x and self.y == other.y

v1 = Vector(1, 2)
v2 = Vector(3, 4)
print(f"{v1} + {v2} = {v1 + v2}")
print(f"v1 == v2: {v1 == v2}")
```

## Properties and Encapsulation

```python
class Circle:
    def __init__(self, radius):
        self._radius = radius  # Convention: _ means "private"
    
    @property
    def radius(self):
        """Getter for radius."""
        return self._radius
    
    @radius.setter
    def radius(self, value):
        """Setter with validation."""
        if value < 0:
            raise ValueError("Radius must be positive")
        self._radius = value
    
    @property
    def area(self):
        """Computed property."""
        import math
        return math.pi * self._radius ** 2

c = Circle(5)
print(f"Radius: {c.radius}, Area: {c.area:.2f}")
c.radius = 10
print(f"New area: {c.area:.2f}")
```

## Dataclasses (Python 3.7+)

Simplified class definition for data containers.

```python
from dataclasses import dataclass

@dataclass
class Point:
    x: float
    y: float
    
    def distance_from_origin(self):
        return (self.x**2 + self.y**2) ** 0.5

p1 = Point(3, 4)
p2 = Point(3, 4)

print(p1)  # Auto-generated __repr__
print(f"p1 == p2: {p1 == p2}")  # Auto-generated __eq__
print(f"Distance: {p1.distance_from_origin()}")
```

## Key Takeaways

| Concept | Key Point |
|---------|----------|
| **`self`** | Reference to current instance |
| **`__init__`** | Constructor, called on instantiation |
| **Inheritance** | `class Child(Parent)` |
| **`@property`** | Computed attributes with getters/setters |
| **`@dataclass`** | Auto-generate boilerplate for data classes |
