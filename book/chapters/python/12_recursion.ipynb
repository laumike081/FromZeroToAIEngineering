# Recursion

A function that calls itself to solve smaller subproblems.

---

## Anatomy of Recursion

Every recursive function needs:
1. **Base case** - when to stop
2. **Recursive case** - break problem into smaller subproblems

```python
def factorial(n):
    """Classic example: n! = n * (n-1)!"""
    # Base case
    if n <= 1:
        return 1
    # Recursive case
    return n * factorial(n - 1)

print(f"5! = {factorial(5)}")
print(f"10! = {factorial(10)}")
```

```python
def fibonacci(n):
    """Fibonacci: F(n) = F(n-1) + F(n-2)"""
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)

print(f"Fib(10) = {fibonacci(10)}")
# Note: This is O(2^n) - very slow!
```

---

## Recursion with Memoization

Cache results to avoid redundant computation.

```python
from functools import lru_cache

@lru_cache(maxsize=None)
def fibonacci_memo(n):
    """Memoized Fibonacci: O(n)"""
    if n <= 1:
        return n
    return fibonacci_memo(n - 1) + fibonacci_memo(n - 2)

print(f"Fib(50) = {fibonacci_memo(50)}")
print(f"Fib(100) = {fibonacci_memo(100)}")
```

## Common Recursive Patterns

### 1. Linear Recursion

```python
def sum_list(arr):
    """Sum elements recursively."""
    if not arr:
        return 0
    return arr[0] + sum_list(arr[1:])

print(f"Sum [1,2,3,4,5] = {sum_list([1,2,3,4,5])}")
```

### 2. Binary Recursion (Divide and Conquer)

```python
def binary_sum(arr, left, right):
    """Sum using divide and conquer."""
    if left > right:
        return 0
    if left == right:
        return arr[left]
    
    mid = (left + right) // 2
    return binary_sum(arr, left, mid) + binary_sum(arr, mid + 1, right)

arr = [1, 2, 3, 4, 5]
print(f"Binary sum = {binary_sum(arr, 0, len(arr) - 1)}")
```

### 3. Backtracking

```python
def permutations(nums):
    """Generate all permutations using backtracking."""
    result = []
    
    def backtrack(path, remaining):
        if not remaining:
            result.append(path[:])
            return
        
        for i in range(len(remaining)):
            path.append(remaining[i])
            backtrack(path, remaining[:i] + remaining[i+1:])
            path.pop()  # Backtrack
    
    backtrack([], nums)
    return result

print(f"Permutations of [1,2,3]:")
for p in permutations([1, 2, 3]):
    print(f"  {p}")
```

```python
def subsets(nums):
    """Generate all subsets."""
    result = []
    
    def backtrack(start, path):
        result.append(path[:])
        
        for i in range(start, len(nums)):
            path.append(nums[i])
            backtrack(i + 1, path)
            path.pop()
    
    backtrack(0, [])
    return result

print(f"Subsets of [1,2,3]: {subsets([1, 2, 3])}")
```

## Recursion vs Iteration

| Aspect | Recursion | Iteration |
|--------|-----------|----------|
| Readability | Often cleaner | Can be verbose |
| Space | O(n) stack | O(1) possible |
| Speed | Function call overhead | Generally faster |
| Stack overflow | Risk with deep recursion | No risk |

```python
# Iterative factorial (more efficient)
def factorial_iter(n):
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result

print(f"Iterative 10! = {factorial_iter(10)}")
```

## Key Takeaways

| Concept | Key Point |
|---------|----------|
| **Base case** | Always define stopping condition |
| **Memoization** | Use `@lru_cache` for overlapping subproblems |
| **Backtracking** | Explore, then undo (pop) |
| **Stack overflow** | Consider iteration for deep recursion |
