# Control Flow

Control flow determines the order in which code executes.

---

## Conditional Statements

```python
x = 10

# if-elif-else
if x > 0:
    print("Positive")
elif x < 0:
    print("Negative")
else:
    print("Zero")
```

```python
# Ternary operator
age = 20
status = "adult" if age >= 18 else "minor"
print(f"Status: {status}")
```

---

## Comparison and Logical Operators

| Operator | Description |
|----------|-------------|
| `==`, `!=` | Equal, not equal |
| `<`, `>`, `<=`, `>=` | Comparisons |
| `and`, `or`, `not` | Logical operators |
| `is`, `is not` | Identity (same object) |
| `in`, `not in` | Membership |

```python
a, b = 5, 10

print(f"a == b: {a == b}")
print(f"a < b and b < 20: {a < b and b < 20}")
print(f"a in [1, 5, 10]: {a in [1, 5, 10]}")

# Chained comparisons
x = 15
print(f"10 < x < 20: {10 < x < 20}")
```

## For Loops

```python
# Iterate over a list
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)

print("---")

# With index using enumerate
for i, fruit in enumerate(fruits):
    print(f"{i}: {fruit}")
```

```python
# range() function
print("range(5):", list(range(5)))           # 0 to 4
print("range(2, 8):", list(range(2, 8)))     # 2 to 7
print("range(0, 10, 2):", list(range(0, 10, 2)))  # Step by 2
```

---

## While Loops

```python
count = 0
while count < 5:
    print(f"Count: {count}")
    count += 1
```

---

## Break, Continue, Pass

```python
# break - exit loop
for i in range(10):
    if i == 5:
        break
    print(i, end=" ")
print("\n---")

# continue - skip iteration
for i in range(10):
    if i % 2 == 0:
        continue
    print(i, end=" ")
print("\n---")

# pass - placeholder (do nothing)
for i in range(3):
    pass  # TODO: implement later
```

## List Comprehensions

Concise way to create lists.

```python
# Basic comprehension
squares = [x**2 for x in range(10)]
print(f"Squares: {squares}")

# With condition
evens = [x for x in range(20) if x % 2 == 0]
print(f"Evens: {evens}")

# Nested comprehension
matrix = [[i*j for j in range(1, 4)] for i in range(1, 4)]
print(f"Matrix: {matrix}")
```
