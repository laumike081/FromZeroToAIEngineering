# Built-in Data Structures

Python's core data structures: lists, tuples, dictionaries, and sets.

---

## Overview

| Type | Ordered | Mutable | Duplicates | Use Case |
|------|---------|---------|------------|----------|
| `list` | ✅ | ✅ | ✅ | General-purpose sequence |
| `tuple` | ✅ | ❌ | ✅ | Immutable sequence, dict keys |
| `dict` | ✅* | ✅ | Keys: ❌ | Key-value mapping |
| `set` | ❌ | ✅ | ❌ | Unique elements, membership |

*Ordered since Python 3.7

---

## Lists

```python
# Creating lists
nums = [1, 2, 3, 4, 5]
mixed = [1, "hello", 3.14, True]

# Indexing and slicing
print(f"First: {nums[0]}, Last: {nums[-1]}")
print(f"Slice [1:4]: {nums[1:4]}")
print(f"Reverse: {nums[::-1]}")
```

```python
# List methods
nums = [3, 1, 4, 1, 5]

nums.append(9)        # Add to end
nums.insert(0, 0)     # Insert at index
nums.extend([2, 6])   # Add multiple
print(f"After adding: {nums}")

nums.remove(1)        # Remove first occurrence
popped = nums.pop()   # Remove and return last
print(f"After removing: {nums}, popped: {popped}")

nums.sort()           # Sort in place
print(f"Sorted: {nums}")
```

### List Time Complexity

| Operation | Time |
|-----------|------|
| `append()` | O(1) |
| `pop()` | O(1) |
| `pop(i)` | O(n) |
| `insert(i, x)` | O(n) |
| `remove(x)` | O(n) |
| `x in list` | O(n) |
| `list[i]` | O(1) |

## Tuples

Immutable sequences - can't be modified after creation.

```python
# Creating tuples
point = (3, 4)
single = (42,)  # Note the comma for single element

# Tuple unpacking
x, y = point
print(f"x={x}, y={y}")

# Swap values
a, b = 1, 2
a, b = b, a
print(f"Swapped: a={a}, b={b}")

# Tuples as dict keys (lists can't be)
locations = {(0, 0): "origin", (1, 0): "east"}
print(locations[(0, 0)])
```

---

## Dictionaries

Key-value pairs with O(1) average lookup.

```python
# Creating dictionaries
person = {"name": "Alice", "age": 30, "city": "NYC"}

# Access and modify
print(f"Name: {person['name']}")
print(f"Age (safe): {person.get('age', 'N/A')}")

person["email"] = "alice@example.com"  # Add
person["age"] = 31                      # Update
del person["city"]                      # Delete

print(f"Updated: {person}")
```

```python
# Iterating dictionaries
scores = {"Alice": 95, "Bob": 87, "Charlie": 92}

for name in scores:           # Keys
    print(name)

for name, score in scores.items():  # Key-value pairs
    print(f"{name}: {score}")
```

```python
# Dictionary comprehension
squares = {x: x**2 for x in range(6)}
print(f"Squares: {squares}")

# defaultdict for counting
from collections import defaultdict, Counter

word_count = defaultdict(int)
for word in ["a", "b", "a", "c", "a", "b"]:
    word_count[word] += 1
print(f"Word count: {dict(word_count)}")

# Counter (even simpler)
counts = Counter(["a", "b", "a", "c", "a", "b"])
print(f"Counter: {counts}")
print(f"Most common: {counts.most_common(2)}")
```

### Dict Time Complexity

| Operation | Average | Worst |
|-----------|---------|-------|
| `d[key]` | O(1) | O(n) |
| `d[key] = val` | O(1) | O(n) |
| `key in d` | O(1) | O(n) |
| `del d[key]` | O(1) | O(n) |

## Sets

Unordered collection of unique elements.

```python
# Creating sets
nums = {1, 2, 3, 2, 1}  # Duplicates removed
print(f"Set: {nums}")

# Set operations
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print(f"Union: {a | b}")
print(f"Intersection: {a & b}")
print(f"Difference: {a - b}")
print(f"Symmetric diff: {a ^ b}")
```

```python
# Fast membership testing
valid_ids = {1001, 1002, 1003, 1004, 1005}

print(f"1003 in valid_ids: {1003 in valid_ids}")  # O(1)
print(f"9999 in valid_ids: {9999 in valid_ids}")
```

### Set Time Complexity

| Operation | Average |
|-----------|--------|
| `x in s` | O(1) |
| `s.add(x)` | O(1) |
| `s.remove(x)` | O(1) |
| `s \| t` (union) | O(len(s) + len(t)) |

---

## Key Takeaways

| Structure | Best For |
|-----------|----------|
| **List** | Ordered collection, frequent access by index |
| **Tuple** | Immutable data, dict keys, function returns |
| **Dict** | Key-value lookup, counting, caching |
| **Set** | Uniqueness, membership testing, set operations |
