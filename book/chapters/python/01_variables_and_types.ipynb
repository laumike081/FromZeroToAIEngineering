# Variables and Data Types

Python is dynamically typed - variables don't need explicit type declarations.

---

## Basic Data Types

| Type | Example | Description |
|------|---------|-------------|
| `int` | `42` | Integers (unlimited precision) |
| `float` | `3.14` | Floating-point numbers |
| `str` | `"hello"` | Strings (immutable) |
| `bool` | `True` | Boolean values |
| `None` | `None` | Null/absence of value |

```python
# Variable assignment
x = 42          # int
pi = 3.14159    # float
name = "Alice"  # str
is_valid = True # bool
nothing = None  # NoneType

print(f"x = {x}, type: {type(x)}")
print(f"pi = {pi}, type: {type(pi)}")
print(f"name = {name}, type: {type(name)}")
```

---

## Type Conversion

```python
# Explicit conversion
a = int("42")      # str -> int
b = float(42)      # int -> float
c = str(3.14)      # float -> str
d = bool(1)        # int -> bool (0 is False, non-zero is True)

print(f"int('42') = {a}")
print(f"float(42) = {b}")
print(f"str(3.14) = {c}")
print(f"bool(1) = {d}, bool(0) = {bool(0)}")
```

---

## Strings

Strings are **immutable** sequences of characters.

```python
s = "Hello, World!"

# Indexing and slicing
print(f"First char: {s[0]}")
print(f"Last char: {s[-1]}")
print(f"Slice [0:5]: {s[0:5]}")

# Common methods
print(f"Upper: {s.upper()}")
print(f"Lower: {s.lower()}")
print(f"Split: {s.split(', ')}")
print(f"Replace: {s.replace('World', 'Python')}")
print(f"Length: {len(s)}")
```

---

## F-Strings (Formatted String Literals)

Python 3.6+ feature for string formatting.

```python
name = "Alice"
age = 30
score = 95.5

# F-string formatting
print(f"Name: {name}, Age: {age}")
print(f"Score: {score:.2f}")  # 2 decimal places
print(f"Binary: {age:b}")     # Binary representation
print(f"Padded: {age:05d}")   # Zero-padded
```

---

## Numbers and Operators

```python
# Arithmetic operators
a, b = 17, 5

print(f"{a} + {b} = {a + b}")   # Addition
print(f"{a} - {b} = {a - b}")   # Subtraction
print(f"{a} * {b} = {a * b}")   # Multiplication
print(f"{a} / {b} = {a / b}")   # Division (float)
print(f"{a} // {b} = {a // b}") # Floor division
print(f"{a} % {b} = {a % b}")   # Modulo
print(f"{a} ** {b} = {a ** b}") # Exponentiation
```

---

## Key Takeaways

| Concept | Key Point |
|---------|----------|
| **Dynamic typing** | Variables infer type from value |
| **Immutable strings** | String operations return new strings |
| **F-strings** | Preferred way to format strings |
| **`//` vs `/`** | Floor division vs true division |
