# Functions

Functions are reusable blocks of code that perform a specific task.

---

## Defining Functions

```python
def greet(name):
    """Return a greeting message."""
    return f"Hello, {name}!"

print(greet("Alice"))
print(greet.__doc__)  # Access docstring
```

---

## Parameters and Arguments

| Type | Description |
|------|-------------|
| Positional | Required, order matters |
| Default | Has default value |
| `*args` | Variable positional arguments |
| `**kwargs` | Variable keyword arguments |

```python
# Default parameters
def power(base, exponent=2):
    return base ** exponent

print(f"power(3) = {power(3)}")
print(f"power(2, 10) = {power(2, 10)}")
```

```python
# *args - variable positional arguments
def sum_all(*args):
    return sum(args)

print(f"sum_all(1, 2, 3) = {sum_all(1, 2, 3)}")
print(f"sum_all(1, 2, 3, 4, 5) = {sum_all(1, 2, 3, 4, 5)}")
```

```python
# **kwargs - variable keyword arguments
def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name="Alice", age=30, city="NYC")
```

## Return Values

```python
# Multiple return values (tuple unpacking)
def min_max(numbers):
    return min(numbers), max(numbers)

lo, hi = min_max([3, 1, 4, 1, 5, 9])
print(f"Min: {lo}, Max: {hi}")
```

---

## Lambda Functions

Anonymous, single-expression functions.

```python
# Lambda syntax: lambda arguments: expression
square = lambda x: x ** 2
print(f"square(5) = {square(5)}")

# Common use: sorting with custom key
points = [(1, 2), (3, 1), (2, 4)]
sorted_by_y = sorted(points, key=lambda p: p[1])
print(f"Sorted by y: {sorted_by_y}")
```

## Scope: Local vs Global

```python
global_var = "I'm global"

def demo_scope():
    local_var = "I'm local"
    print(global_var)  # Can read global
    print(local_var)

demo_scope()
# print(local_var)  # Error: local_var not defined outside
```

```python
# Modifying global variables
counter = 0

def increment():
    global counter
    counter += 1

increment()
increment()
print(f"Counter: {counter}")
```

## Type Hints (Python 3.5+)

Optional type annotations for better code documentation.

```python
def add(a: int, b: int) -> int:
    """Add two integers."""
    return a + b

def process_items(items: list[str]) -> dict[str, int]:
    """Count items."""
    return {item: len(item) for item in items}

print(add(3, 5))
print(process_items(["apple", "banana"]))
```
