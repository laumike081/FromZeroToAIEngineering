# Stacks and Queues

Linear data structures with restricted access patterns.

---

## Stack (LIFO - Last In, First Out)

Think of a stack of plates - you can only add/remove from the top.

| Operation | Description | Time |
|-----------|-------------|------|
| `push(x)` | Add to top | O(1) |
| `pop()` | Remove from top | O(1) |
| `peek()` | View top | O(1) |

```python
# Stack using list
stack = []

# Push
stack.append(1)
stack.append(2)
stack.append(3)
print(f"Stack: {stack}")

# Peek
print(f"Top: {stack[-1]}")

# Pop
print(f"Popped: {stack.pop()}")
print(f"Stack: {stack}")
```

```python
# Example: Balanced parentheses
def is_balanced(s: str) -> bool:
    stack = []
    pairs = {')': '(', ']': '[', '}': '{'}
    
    for char in s:
        if char in '([{':
            stack.append(char)
        elif char in ')]}':
            if not stack or stack.pop() != pairs[char]:
                return False
    
    return len(stack) == 0

print(f"'()[]{{}}': {is_balanced('()[]{}')}")  # True
print(f"'([)]': {is_balanced('([)]')}")        # False
print(f"'{{[]}}': {is_balanced('{[]}')}")
```

## Queue (FIFO - First In, First Out)

Think of a line at a store - first person in line is served first.

| Operation | Description | Time |
|-----------|-------------|------|
| `enqueue(x)` | Add to back | O(1) |
| `dequeue()` | Remove from front | O(1) |
| `peek()` | View front | O(1) |

```python
from collections import deque

# Queue using deque (efficient)
queue = deque()

# Enqueue
queue.append(1)
queue.append(2)
queue.append(3)
print(f"Queue: {list(queue)}")

# Peek
print(f"Front: {queue[0]}")

# Dequeue
print(f"Dequeued: {queue.popleft()}")
print(f"Queue: {list(queue)}")
```

### Why `deque` instead of `list`?

| Operation | `list` | `deque` |
|-----------|--------|--------|
| `append()` | O(1) | O(1) |
| `pop()` | O(1) | O(1) |
| `pop(0)` / `popleft()` | **O(n)** | **O(1)** |
| `insert(0, x)` / `appendleft()` | **O(n)** | **O(1)** |

```python
# Example: BFS using queue
def bfs(graph, start):
    visited = set()
    queue = deque([start])
    result = []
    
    while queue:
        node = queue.popleft()
        if node not in visited:
            visited.add(node)
            result.append(node)
            queue.extend(graph.get(node, []))
    
    return result

graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [], 'E': [], 'F': []
}
print(f"BFS: {bfs(graph, 'A')}")
```

## Deque (Double-Ended Queue)

Efficient operations at both ends.

```python
from collections import deque

d = deque([1, 2, 3])

d.append(4)       # Add right
d.appendleft(0)   # Add left
print(f"After appends: {list(d)}")

d.pop()           # Remove right
d.popleft()       # Remove left
print(f"After pops: {list(d)}")

# Rotate
d = deque([1, 2, 3, 4, 5])
d.rotate(2)       # Rotate right by 2
print(f"Rotated right: {list(d)}")
d.rotate(-2)      # Rotate left by 2
print(f"Rotated back: {list(d)}")
```

## Priority Queue (Heap)

Elements are dequeued by priority, not insertion order.

```python
import heapq

# Min-heap (smallest first)
heap = []
heapq.heappush(heap, 3)
heapq.heappush(heap, 1)
heapq.heappush(heap, 4)
heapq.heappush(heap, 1)
heapq.heappush(heap, 5)

print(f"Heap: {heap}")
print(f"Pop min: {heapq.heappop(heap)}")
print(f"Pop min: {heapq.heappop(heap)}")

# Get k smallest/largest
nums = [3, 1, 4, 1, 5, 9, 2, 6]
print(f"3 smallest: {heapq.nsmallest(3, nums)}")
print(f"3 largest: {heapq.nlargest(3, nums)}")
```

### Heap Time Complexity

| Operation | Time |
|-----------|------|
| `heappush()` | O(log n) |
| `heappop()` | O(log n) |
| `heapify()` | O(n) |
| `heap[0]` (peek min) | O(1) |

## Key Takeaways

| Structure | Pattern | Use Case |
|-----------|---------|----------|
| **Stack** | LIFO | Undo, DFS, parsing |
| **Queue** | FIFO | BFS, task scheduling |
| **Deque** | Both ends | Sliding window |
| **Heap** | Priority | Top-K, Dijkstra |
