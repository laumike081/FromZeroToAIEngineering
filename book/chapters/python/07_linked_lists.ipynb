# Linked Lists

A linear data structure where elements are stored in nodes, each pointing to the next.

---

## Singly Linked List

Each node has data and a pointer to the next node.

```
[1] -> [2] -> [3] -> [4] -> None
```

````python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
    
    def __repr__(self):
        return f"ListNode({self.val})"

# Create linked list: 1 -> 2 -> 3
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)

# Traverse and print
def print_list(head):
    vals = []
    while head:
        vals.append(str(head.val))
        head = head.next
    return " -> ".join(vals)

print(print_list(head))
````

### List vs Linked List

| Operation | Array/List | Linked List |
|-----------|------------|-------------|
| Access by index | O(1) | O(n) |
| Insert at beginning | O(n) | O(1) |
| Insert at end | O(1)* | O(n)** |
| Delete at beginning | O(n) | O(1) |
| Search | O(n) | O(n) |

*Amortized, **O(1) with tail pointer

````python
# Common operations

def insert_at_head(head, val):
    """Insert at beginning: O(1)"""
    new_node = ListNode(val)
    new_node.next = head
    return new_node

def delete_node(head, val):
    """Delete first occurrence: O(n)"""
    dummy = ListNode(0, head)
    prev = dummy
    
    while prev.next:
        if prev.next.val == val:
            prev.next = prev.next.next
            break
        prev = prev.next
    
    return dummy.next

# Test
head = ListNode(1, ListNode(2, ListNode(3)))
head = insert_at_head(head, 0)
print(f"After insert: {print_list(head)}")

head = delete_node(head, 2)
print(f"After delete: {print_list(head)}")
````

## Common Patterns

### 1. Two Pointers (Slow/Fast)

````python
def find_middle(head):
    """Find middle node using slow/fast pointers."""
    slow = fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
    return slow

def has_cycle(head):
    """Detect cycle using Floyd's algorithm."""
    slow = fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            return True
    return False

# Test find middle
head = ListNode(1, ListNode(2, ListNode(3, ListNode(4, ListNode(5)))))
print(f"Middle of 1->2->3->4->5: {find_middle(head).val}")
````

### 2. Reverse Linked List

````python
def reverse_list(head):
    """Reverse linked list iteratively: O(n)"""
    prev = None
    curr = head
    
    while curr:
        next_node = curr.next  # Save next
        curr.next = prev       # Reverse pointer
        prev = curr            # Move prev forward
        curr = next_node       # Move curr forward
    
    return prev

# Test
head = ListNode(1, ListNode(2, ListNode(3, ListNode(4))))
print(f"Original: {print_list(head)}")
reversed_head = reverse_list(head)
print(f"Reversed: {print_list(reversed_head)}")
````

### 3. Dummy Head Pattern

````python
def merge_two_lists(l1, l2):
    """Merge two sorted lists using dummy head."""
    dummy = ListNode(0)
    curr = dummy
    
    while l1 and l2:
        if l1.val <= l2.val:
            curr.next = l1
            l1 = l1.next
        else:
            curr.next = l2
            l2 = l2.next
        curr = curr.next
    
    curr.next = l1 or l2
    return dummy.next

# Test
l1 = ListNode(1, ListNode(3, ListNode(5)))
l2 = ListNode(2, ListNode(4, ListNode(6)))
merged = merge_two_lists(l1, l2)
print(f"Merged: {print_list(merged)}")
````

## Key Takeaways

| Pattern | Use Case |
|---------|----------|
| **Slow/Fast pointers** | Find middle, detect cycle |
| **Dummy head** | Simplify edge cases |
| **Reverse** | Many problems require partial reversal |
| **prev/curr/next** | Standard traversal pattern |
