## Problem Statement

Given an array `nums`. We define a running sum of an array as `runningSum[i] = sum(nums[0]…nums[i])`.

Return the running sum of `nums`.

### Examples

**Example 1:**
```
Input: nums = [1,2,3,4]
Output: [1,3,6,10]
```

### Constraints

- `1 <= nums.length <= 1000`
- `-10^6 <= nums[i] <= 10^6`

---

## Approach: Prefix Sum

### Complexity
- **Time**: O(n)
- **Space**: O(1) extra

````python
def running_sum(nums: list[int]) -> list[int]:
    """In-place prefix sum: O(n)"""
    for i in range(1, len(nums)):
        nums[i] += nums[i - 1]
    return nums

# Test
print(running_sum([1, 2, 3, 4]))  # [1, 3, 6, 10]
````

---

## Using itertools

````python
from itertools import accumulate

def running_sum_accumulate(nums: list[int]) -> list[int]:
    return list(accumulate(nums))

print(running_sum_accumulate([1, 2, 3, 4]))  # [1, 3, 6, 10]
````
