## Problem Statement

Write a function that reverses a string. The input string is given as an array of characters `s`.

You must do this by modifying the input array **in-place** with O(1) extra memory.

### Examples

**Example 1:**
```
Input: s = ["h","e","l","l","o"]
Output: ["o","l","l","e","h"]
```

**Example 2:**
```
Input: s = ["H","a","n","n","a","h"]
Output: ["h","a","n","n","a","H"]
```

### Constraints

- `1 <= s.length <= 10^5`
- `s[i]` is a printable ASCII character

---

## Approach: Two Pointers

### Intuition

Use two pointers at both ends, swap and move towards center.

### Complexity
- **Time**: O(n)
- **Space**: O(1)

````python
def reverse_string(s: list[str]) -> None:
    """Two pointers: O(n) time, O(1) space"""
    left, right = 0, len(s) - 1
    
    while left < right:
        s[left], s[right] = s[right], s[left]
        left += 1
        right -= 1

# Test
s = ["h", "e", "l", "l", "o"]
reverse_string(s)
print(s)  # ['o', 'l', 'l', 'e', 'h']
````

## Walkthrough

Input: `["h", "e", "l", "l", "o"]`

| Step | left | right | Swap | Array |
|------|------|-------|------|-------|
| 0 | 0 | 4 | h <-> o | [o,e,l,l,h] |
| 1 | 1 | 3 | e <-> l | [o,l,l,e,h] |
| 2 | 2 | 2 | done | [o,l,l,e,h] |

---

## Alternative: Recursion

````python
def reverse_string_recursive(s: list[str]) -> None:
    """Recursive approach (not O(1) space due to call stack)"""
    def helper(left, right):
        if left >= right:
            return
        s[left], s[right] = s[right], s[left]
        helper(left + 1, right - 1)
    
    helper(0, len(s) - 1)

s = ["h", "e", "l", "l", "o"]
reverse_string_recursive(s)
print(s)  # ['o', 'l', 'l', 'e', 'h']
````

## Key Takeaways

| Pattern | When to Use |
|---------|-------------|
| **Two pointers from ends** | In-place array reversal |
| **Swap until pointers meet** | Symmetric operations |
