## Problem Statement

Reverse bits of a given 32 bits unsigned integer.

### Examples

**Example 1:**
```
Input: n = 00000010100101000001111010011100
Output:    964176192 (00111001011110000010100101000000)
```

**Example 2:**
```
Input: n = 11111111111111111111111111111101
Output:   3221225471 (10111111111111111111111111111111)
```

### Constraints

- The input must be a **binary string** of length `32`

**Follow up:** If this function is called many times, how would you optimize it?

---

## Approach 1: Bit by Bit

### Intuition

Process each bit from right to left:
1. Extract the last bit of `n`
2. Add it to result (shifted to correct position)
3. Shift `n` right

### Complexity
- **Time**: O(32) = O(1)
- **Space**: O(1)

````python
def reverse_bits(n: int) -> int:
    """Bit by bit: O(32) time"""
    result = 0
    
    for i in range(32):
        # Extract last bit of n
        bit = n & 1
        
        # Add to result at reversed position
        result = (result << 1) | bit
        
        # Shift n right
        n >>= 1
    
    return result

# Test
print(reverse_bits(0b00000010100101000001111010011100))  # 964176192
print(bin(reverse_bits(0b00000010100101000001111010011100)))  # 0b111001011110000010100101000000
````

## Walkthrough

Input: `n = 5` (binary: `101`)

| i | n (binary) | bit | result (binary) |
|---|------------|-----|----------------|
| 0 | 101 | 1 | 1 |
| 1 | 10 | 0 | 10 |
| 2 | 1 | 1 | 101 |
| ... | 0 | 0 | 1010... |

After 32 iterations: `10100000000000000000000000000000`

````python
def reverse_bits_verbose(n: int) -> int:
    """Verbose version"""
    result = 0
    original = n
    
    print(f"Input: {n} = {bin(n)}")
    print(f"32-bit: {n:032b}\n")
    
    for i in range(32):
        bit = n & 1
        result = (result << 1) | bit
        n >>= 1
        
        if i < 5 or i >= 30:  # Show first and last few
            print(f"Step {i:2d}: bit={bit}, result={result:032b}")
        elif i == 5:
            print("...")
    
    print(f"\nResult: {result} = {bin(result)}")
    return result

reverse_bits_verbose(43261596)
````

## Approach 2: Divide and Conquer (Optimized for Multiple Calls)

### Intuition

Swap halves recursively:
1. Swap 16-bit halves
2. Swap 8-bit halves within each 16-bit
3. Continue down to single bits

This is faster for repeated calls due to parallelism.

````python
def reverse_bits_divide_conquer(n: int) -> int:
    """Divide and conquer with masks"""
    # Swap adjacent bits
    n = ((n & 0x55555555) << 1) | ((n & 0xAAAAAAAA) >> 1)
    # Swap adjacent pairs
    n = ((n & 0x33333333) << 2) | ((n & 0xCCCCCCCC) >> 2)
    # Swap adjacent nibbles (4 bits)
    n = ((n & 0x0F0F0F0F) << 4) | ((n & 0xF0F0F0F0) >> 4)
    # Swap adjacent bytes
    n = ((n & 0x00FF00FF) << 8) | ((n & 0xFF00FF00) >> 8)
    # Swap 16-bit halves
    n = ((n & 0x0000FFFF) << 16) | ((n & 0xFFFF0000) >> 16)
    
    return n

# Test
print(reverse_bits_divide_conquer(43261596))  # 964176192
````

## Python One-Liner

````python
def reverse_bits_pythonic(n: int) -> int:
    """Using string manipulation"""
    return int(bin(n)[2:].zfill(32)[::-1], 2)

# Test
print(reverse_bits_pythonic(43261596))  # 964176192
````

## Key Takeaways

| Pattern | When to Use |
|---------|-------------|
