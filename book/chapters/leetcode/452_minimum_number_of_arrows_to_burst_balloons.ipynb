## Problem Statement

There are some spherical balloons taped onto a flat wall that represents the XY-plane. The balloons are represented as a 2D integer array `points` where `points[i] = [xstart, xend]` denotes a balloon whose horizontal diameter stretches between `xstart` and `xend`.

An arrow can be shot up exactly vertically (in the positive y-direction) from different points along the x-axis. A balloon with `xstart` and `xend` is burst by an arrow shot at `x` if `xstart <= x <= xend`.

Return the minimum number of arrows that must be shot to burst all balloons.

### Constraints

- `1 <= points.length <= 10^5`

```python
def find_min_arrow_shots(points: list[list[int]]) -> int:
    """Greedy: sort by end, O(n log n)"""
    points.sort(key=lambda x: x[1])
    
    arrows = 1
    arrow_pos = points[0][1]
    
    for start, end in points[1:]:
        if start > arrow_pos:
            arrows += 1
            arrow_pos = end
    
    return arrows

# Test
print(find_min_arrow_shots([[10,16],[2,8],[1,6],[7,12]]))  # 2
```
