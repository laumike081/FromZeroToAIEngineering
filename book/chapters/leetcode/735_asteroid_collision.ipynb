## Problem Statement

We are given an array `asteroids` of integers representing asteroids in a row.

For each asteroid, the absolute value represents its size, and the sign represents its direction (positive meaning right, negative meaning left). Each asteroid moves at the same speed.

Find out the state of the asteroids after all collisions. If two asteroids meet, the smaller one will explode. If both are the same size, both will explode. Two asteroids moving in the same direction will never meet.

### Examples

**Example 1:**
```
Input: asteroids = [5,10,-5]
Output: [5,10]
Explanation: The 10 and -5 collide, 10 survives. The 5 and 10 never collide.
```

**Example 2:**
```
Input: asteroids = [8,-8]
Output: []
Explanation: The 8 and -8 collide, both explode.
```

**Example 3:**
```
Input: asteroids = [10,2,-5]
Output: [10]
Explanation: The 2 and -5 collide, -5 survives. Then 10 and -5 collide, 10 survives.
```

### Constraints

- `2 <= asteroids.length <= 10^4`
- `-1000 <= asteroids[i] <= 1000`
- `asteroids[i] != 0`

---

## Approach: Stack

### Intuition

Collision only happens when:
- Stack top is moving **right** (positive)
- Current asteroid is moving **left** (negative)

No collision if:
- Both moving same direction
- Stack top moving left, current moving right (they move apart)

### Algorithm

1. For each asteroid:
   - While collision possible (stack top > 0, current < 0):
     - Compare sizes and handle explosion
   - If asteroid survives, push to stack

### Complexity
- **Time**: O(n) - each asteroid pushed/popped at most once
- **Space**: O(n) - stack

````python
def asteroid_collision(asteroids: list[int]) -> list[int]:
    """Stack approach: O(n) time and space"""
    stack = []
    
    for asteroid in asteroids:
        alive = True
        
        # Collision: stack top moving right (+), current moving left (-)
        while alive and stack and stack[-1] > 0 and asteroid < 0:
            # Compare sizes
            if stack[-1] < abs(asteroid):
                # Stack top explodes, continue checking
                stack.pop()
            elif stack[-1] == abs(asteroid):
                # Both explode
                stack.pop()
                alive = False
            else:
                # Current asteroid explodes
                alive = False
        
        if alive:
            stack.append(asteroid)
    
    return stack

# Test
print(asteroid_collision([5, 10, -5]))   # [5, 10]
print(asteroid_collision([8, -8]))       # []
print(asteroid_collision([10, 2, -5]))   # [10]
````

## Walkthrough

Input: `[10, 2, -5]`

| Step | Asteroid | Stack | Collision? | Result |
|------|----------|-------|------------|--------|
| 0 | 10 | [] | No | [10] |
| 1 | 2 | [10] | No (both +) | [10, 2] |
| 2 | -5 | [10, 2] | Yes: 2 vs 5 | 2 explodes |
| 2 | -5 | [10] | Yes: 10 vs 5 | -5 explodes |
| - | - | [10] | - | **[10]** |

````python
def asteroid_collision_verbose(asteroids: list[int]) -> list[int]:
    """Verbose version showing collisions"""
    stack = []
    
    print(f"Input: {asteroids}\n")
    
    for asteroid in asteroids:
        print(f"Processing asteroid {asteroid} (direction: {'right' if asteroid > 0 else 'left'})")
        alive = True
        
        while alive and stack and stack[-1] > 0 and asteroid < 0:
            print(f"  Collision! {stack[-1]} vs {asteroid}")
            
            if stack[-1] < abs(asteroid):
                print(f"    {stack[-1]} explodes")
                stack.pop()
            elif stack[-1] == abs(asteroid):
                print(f"    Both explode")
                stack.pop()
                alive = False
            else:
                print(f"    {asteroid} explodes")
                alive = False
        
        if alive:
            stack.append(asteroid)
            print(f"  Asteroid {asteroid} survives")
        
        print(f"  Stack: {stack}\n")
    
    return stack

asteroid_collision_verbose([10, 2, -5])
````

---

## Edge Cases

````python
# All moving same direction - no collision
print(asteroid_collision([1, 2, 3]))     # [1, 2, 3]
print(asteroid_collision([-1, -2, -3]))  # [-1, -2, -3]

# Moving apart - no collision
print(asteroid_collision([-2, 1]))       # [-2, 1]

# Chain reaction
print(asteroid_collision([5, 10, -15]))  # [-15]
````

---

## Common Mistakes

1. **Wrong collision condition** - only `stack[-1] > 0` AND `asteroid < 0`
2. **Forgetting chain reactions** - one asteroid can destroy multiple
3. **Not handling equal size** - both should explode
4. **Using size instead of abs()** - asteroid value is signed

## Key Takeaways

| Pattern | When to Use |
|---------|-------------|
| **Stack for collision simulation** | Objects moving in opposite directions |
| **While loop for chain reactions** | One event can trigger multiple |
| **Sign = direction, abs = size** | Common pattern in physics problems |
