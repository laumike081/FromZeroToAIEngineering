## Problem Statement

Given 3 positives numbers `a`, `b` and `c`. Return the minimum flips required in some bits of `a` and `b` to make (`a` OR `b` == `c`). (bitwise OR operation).

Flip operation consists of changing any single bit 1 to 0 or changing the bit 0 to 1 in their binary representation.

### Constraints

- `1 <= a <= 10^9`
- `1 <= b <= 10^9`
- `1 <= c <= 10^9`

```python
def min_flips(a: int, b: int, c: int) -> int:
    """Bit by bit comparison: O(32)"""
    flips = 0
    
    while a or b or c:
        bit_a = a & 1
        bit_b = b & 1
        bit_c = c & 1
        
        if bit_c == 0:
            # Both a and b bits must be 0
            flips += bit_a + bit_b
        else:
            # At least one of a or b must be 1
            if bit_a == 0 and bit_b == 0:
                flips += 1
        
        a >>= 1
        b >>= 1
        c >>= 1
    
    return flips

# Test
print(min_flips(2, 6, 5))  # 3
print(min_flips(4, 2, 7))  # 1
```
