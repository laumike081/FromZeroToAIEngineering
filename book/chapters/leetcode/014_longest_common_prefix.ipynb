# 14. Longest Common Prefix

**Difficulty**: Easy  
**Topics**: String, Trie  
**Link**: [LeetCode 14](https://leetcode.com/problems/longest-common-prefix/)

---

## Problem Statement

Write a function to find the longest common prefix string amongst an array of strings.

If there is no common prefix, return an empty string `""`.

### Examples

**Example 1:**
```
Input: strs = ["flower","flow","flight"]
Output: "fl"
```

**Example 2:**
```
Input: strs = ["dog","racecar","car"]
Output: ""
Explanation: There is no common prefix among the input strings.
```

### Constraints

- `1 <= strs.length <= 200`
- `0 <= strs[i].length <= 200`
- `strs[i]` consists of only lowercase English letters.

---

## Approach 1: Vertical Scanning

### Intuition

Compare characters column by column (vertically) across all strings:
1. Check the i-th character of each string
2. If all match, continue to i+1
3. If any mismatch or string ends, return prefix so far

### Complexity
- **Time**: O(S) where S = sum of all characters in all strings
- **Space**: O(1)

````python
def longest_common_prefix(strs: list[str]) -> str:
    """Vertical scanning: O(S) time, O(1) space"""
    if not strs:
        return ""
    
    # Use first string as reference
    for i in range(len(strs[0])):
        char = strs[0][i]
        
        # Check this character against all other strings
        for s in strs[1:]:
            # If we've reached end of string or character doesn't match
            if i >= len(s) or s[i] != char:
                return strs[0][:i]
    
    # First string is the common prefix
    return strs[0]

# Test
print(longest_common_prefix(["flower", "flow", "flight"]))  # "fl"
print(longest_common_prefix(["dog", "racecar", "car"]))     # ""
print(longest_common_prefix(["a"]))                          # "a"
print(longest_common_prefix([""]))                           # ""
````

## Approach 2: Horizontal Scanning

### Intuition

Start with first string as prefix, then shrink it by comparing with each subsequent string.

```
LCP(S1, S2, ..., Sn) = LCP(LCP(LCP(S1, S2), S3), ..., Sn)
```

### Complexity
- **Time**: O(S)
- **Space**: O(1)

````python
def longest_common_prefix_horizontal(strs: list[str]) -> str:
    """Horizontal scanning"""
    if not strs:
        return ""
    
    prefix = strs[0]
    
    for s in strs[1:]:
        # Shrink prefix until it matches start of s
        while not s.startswith(prefix):
            prefix = prefix[:-1]  # Remove last character
            if not prefix:
                return ""
    
    return prefix

# Test
print(longest_common_prefix_horizontal(["flower", "flow", "flight"]))  # "fl"
print(longest_common_prefix_horizontal(["dog", "racecar", "car"]))     # ""
````

---

## Approach 3: Sort and Compare First/Last

### Intuition

After sorting lexicographically:
- First and last strings are most different
- Common prefix of first and last = common prefix of all

### Complexity
- **Time**: O(n log n * m) for sorting, where n = number of strings, m = avg length
- **Space**: O(1) or O(n) depending on sort implementation

````python
def longest_common_prefix_sort(strs: list[str]) -> str:
    """Sort and compare first/last"""
    if not strs:
        return ""
    
    strs.sort()
    first, last = strs[0], strs[-1]
    
    # Find common prefix between first and last
    i = 0
    while i < len(first) and i < len(last) and first[i] == last[i]:
        i += 1
    
    return first[:i]

# Test
print(longest_common_prefix_sort(["flower", "flow", "flight"]))  # "fl"
print(longest_common_prefix_sort(["dog", "racecar", "car"]))     # ""
````

## Approach 4: Using Python's zip

### Intuition

Use `zip(*strs)` to iterate through characters at same positions across all strings.

````python
def longest_common_prefix_zip(strs: list[str]) -> str:
    """Pythonic approach using zip"""
    if not strs:
        return ""
    
    prefix = []
    
    # zip stops at shortest string
    for chars in zip(*strs):
        # chars is tuple of characters at same position
        if len(set(chars)) == 1:  # All same
            prefix.append(chars[0])
        else:
            break
    
    return "".join(prefix)

# Test
print(longest_common_prefix_zip(["flower", "flow", "flight"]))  # "fl"
print(longest_common_prefix_zip(["dog", "racecar", "car"]))     # ""
````

## Walkthrough

Input: `["flower", "flow", "flight"]`

### Vertical Scanning:

| i | char | flower | flow | flight | Match? |
|---|------|--------|------|--------|--------|
| 0 | 'f' | f | f | f | Yes |
| 1 | 'l' | l | l | l | Yes |
| 2 | 'o' | o | o | i | **No** |

Return `"fl"` (first 2 characters)

````python
def longest_common_prefix_verbose(strs: list[str]) -> str:
    """Verbose version showing the process"""
    if not strs:
        return ""
    
    print(f"Input: {strs}")
    print(f"Reference string: '{strs[0]}'")
    print()
    
    for i in range(len(strs[0])):
        char = strs[0][i]
        print(f"Position {i}: checking '{char}'")
        
        for j, s in enumerate(strs[1:], 1):
            if i >= len(s):
                print(f"  String '{s}' too short, return '{strs[0][:i]}'")
                return strs[0][:i]
            
            if s[i] != char:
                print(f"  String '{s}' has '{s[i]}' != '{char}', return '{strs[0][:i]}'")
                return strs[0][:i]
            
            print(f"  String '{s}' has '{s[i]}' - match!")
    
    print(f"\nAll characters matched, return '{strs[0]}'")
    return strs[0]

longest_common_prefix_verbose(["flower", "flow", "flight"])
````

## Comparison of Approaches

| Approach | Time | Space | Best For |
|----------|------|-------|----------|
| **Vertical** | O(S) | O(1) | General use |
| **Horizontal** | O(S) | O(1) | Early termination |
| **Sort** | O(n log n * m) | O(1) | Already sorted input |
| **Zip** | O(S) | O(m) | Pythonic, readable |

---

## Common Mistakes

1. **Not handling empty array** - check `if not strs`
2. **Not handling empty strings** - `[""]` should return `""`
3. **Index out of bounds** - check string length before accessing
4. **Returning wrong slice** - `strs[0][:i]` not `strs[0][:i+1]`
