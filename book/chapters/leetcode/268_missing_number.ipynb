# 268. Missing Number

**Difficulty**: Easy  
**Topics**: Array, Hash Table, Math, Bit Manipulation  
**Link**: [LeetCode 268](https://leetcode.com/problems/missing-number/)

---

## Problem Statement

Given an array `nums` containing `n` distinct numbers in the range `[0, n]`, return the only number in the range that is missing from the array.

### Examples

**Example 1:**
```
Input: nums = [3,0,1]
Output: 2
Explanation: n = 3 since there are 3 numbers, so all numbers are in [0,3]. 2 is missing.
```

**Example 2:**
```
Input: nums = [0,1]
Output: 2
```

**Example 3:**
```
Input: nums = [9,6,4,2,3,5,7,0,1]
Output: 8
```

### Constraints

- `n == nums.length`
- `1 <= n <= 10^4`
- `0 <= nums[i] <= n`
- All numbers are **unique**

---

## Approach 1: Math (Sum Formula)

### Intuition

Expected sum of 0 to n: `n * (n + 1) / 2`

Missing number = Expected sum - Actual sum

### Complexity
- **Time**: O(n)
- **Space**: O(1)

````python
def missing_number(nums: list[int]) -> int:
    """Sum formula: O(n) time, O(1) space"""
    n = len(nums)
    expected_sum = n * (n + 1) // 2
    actual_sum = sum(nums)
    return expected_sum - actual_sum

# Test
print(missing_number([3, 0, 1]))  # 2
print(missing_number([0, 1]))     # 2
print(missing_number([9, 6, 4, 2, 3, 5, 7, 0, 1]))  # 8
````

---

## Approach 2: XOR (Bit Manipulation)

### Intuition

XOR properties:
- `a ^ a = 0`
- `a ^ 0 = a`

XOR all numbers from 0 to n with all numbers in array. Pairs cancel out, leaving the missing number.

### Complexity
- **Time**: O(n)
- **Space**: O(1)

````python
def missing_number_xor(nums: list[int]) -> int:
    """XOR approach: O(n) time, O(1) space"""
    n = len(nums)
    result = n  # Start with n (since indices go 0 to n-1)
    
    for i, num in enumerate(nums):
        result ^= i ^ num
    
    return result

# Test
print(missing_number_xor([3, 0, 1]))  # 2
````

---

## Walkthrough (XOR)

Input: `[3, 0, 1]`, n = 3

```
result = 3
i=0: result = 3 ^ 0 ^ 3 = 0
i=1: result = 0 ^ 1 ^ 0 = 1
i=2: result = 1 ^ 2 ^ 1 = 2
```

**Result**: 2

---

## Approach 3: Hash Set

````python
def missing_number_set(nums: list[int]) -> int:
    """Hash set: O(n) time and space"""
    num_set = set(nums)
    n = len(nums)
    
    for i in range(n + 1):
        if i not in num_set:
            return i
    
    return -1

print(missing_number_set([3, 0, 1]))  # 2
````

## Comparison

| Approach | Time | Space | Notes |
|----------|------|-------|-------|
| **Sum** | O(n) | O(1) | Risk of overflow in some languages |
| **XOR** | O(n) | O(1) | No overflow risk |
| **Set** | O(n) | O(n) | Most intuitive |
