# 1732. Find the Highest Altitude

**Difficulty**: Easy

## Problem Statement

There is a biker going on a road trip. The road trip consists of `n + 1` points at different altitudes. The biker starts his trip on point `0` with altitude equal `0`.

You are given an integer array `gain` of length `n` where `gain[i]` is the **net gain in altitude** between points `i` and `i + 1` for all (`0 <= i < n`). Return the **highest altitude** of a point.

### Examples

**Example 1:**
```
Input: gain = [-5,1,5,0,-7]
Output: 1
Explanation: The altitudes are [0,-5,-4,1,1,-6]. The highest is 1.
```

**Example 2:**
```
Input: gain = [-4,-3,-2,-1,4,3,2]
Output: 0
Explanation: The altitudes are [0,-4,-7,-9,-10,-6,-3,-1]. The highest is 0.
```

### Constraints

- `n == gain.length`
- `1 <= n <= 100`
- `-100 <= gain[i] <= 100`

---

## Approach: Prefix Sum

### Intuition

Altitude at each point is the cumulative sum of gains:
- Start at altitude 0
- Each `gain[i]` changes the altitude
- Track the maximum altitude seen

### Complexity
- **Time**: O(n)
- **Space**: O(1)

````python
def largest_altitude(gain: list[int]) -> int:
    """Prefix sum: O(n) time, O(1) space"""
    altitude = 0
    max_altitude = 0  # Start at 0
    
    for g in gain:
        altitude += g
        max_altitude = max(max_altitude, altitude)
    
    return max_altitude

# Test
print(largest_altitude([-5, 1, 5, 0, -7]))        # 1
print(largest_altitude([-4, -3, -2, -1, 4, 3, 2])) # 0
````

## Walkthrough

Input: `[-5, 1, 5, 0, -7]`

| Point | Gain | Altitude | Max |
|-------|------|----------|-----|
| 0 | - | 0 | 0 |
| 1 | -5 | -5 | 0 |
| 2 | 1 | -4 | 0 |
| 3 | 5 | **1** | **1** |
| 4 | 0 | 1 | 1 |
| 5 | -7 | -6 | 1 |

**Result**: 1

````python
def largest_altitude_verbose(gain: list[int]) -> int:
    """Verbose version"""
    altitude = 0
    max_altitude = 0
    altitudes = [0]
    
    for g in gain:
        altitude += g
        altitudes.append(altitude)
        max_altitude = max(max_altitude, altitude)
    
    print(f"Gains: {gain}")
    print(f"Altitudes: {altitudes}")
    print(f"Highest: {max_altitude}")
    
    return max_altitude

largest_altitude_verbose([-5, 1, 5, 0, -7])
````

## One-Liner with itertools

````python
from itertools import accumulate

def largest_altitude_oneliner(gain: list[int]) -> int:
    return max(accumulate(gain, initial=0))

print(largest_altitude_oneliner([-5, 1, 5, 0, -7]))  # 1
````
