# 219. Contains Duplicate II

**Difficulty**: Easy  
**Topics**: Array, Hash Table, Sliding Window  
**Link**: [LeetCode 219](https://leetcode.com/problems/contains-duplicate-ii/)

---

## Problem Statement

Given an integer array `nums` and an integer `k`, return `true` if there are two **distinct indices** `i` and `j` in the array such that `nums[i] == nums[j]` and `abs(i - j) <= k`.

### Examples

**Example 1:**
```
Input: nums = [1,2,3,1], k = 3
Output: true
```

**Example 2:**
```
Input: nums = [1,0,1,1], k = 1
Output: true
```

**Example 3:**
```
Input: nums = [1,2,3,1,2,3], k = 2
Output: false
```

### Constraints

- `1 <= nums.length <= 10^5`
- `-10^9 <= nums[i] <= 10^9`
- `0 <= k <= 10^5`

---

## Approach 1: Hash Map (Store Last Index)

### Intuition

Store the last seen index of each number. When we see a duplicate, check if distance <= k.

### Complexity
- **Time**: O(n)
- **Space**: O(n)

````python
def contains_nearby_duplicate(nums: list[int], k: int) -> bool:
    """Hash map storing last index: O(n) time and space"""
    last_seen = {}  # value -> last index
    
    for i, num in enumerate(nums):
        if num in last_seen and i - last_seen[num] <= k:
            return True
        last_seen[num] = i
    
    return False

# Test
print(contains_nearby_duplicate([1, 2, 3, 1], 3))        # True
print(contains_nearby_duplicate([1, 0, 1, 1], 1))        # True
print(contains_nearby_duplicate([1, 2, 3, 1, 2, 3], 2))  # False
````

---

## Approach 2: Sliding Window Set

### Intuition

Maintain a set of size k (sliding window). If duplicate found in window, return True.

````python
def contains_nearby_duplicate_window(nums: list[int], k: int) -> bool:
    """Sliding window set: O(n) time, O(k) space"""
    window = set()
    
    for i, num in enumerate(nums):
        if num in window:
            return True
        
        window.add(num)
        
        # Maintain window size k
        if len(window) > k:
            window.remove(nums[i - k])
    
    return False

print(contains_nearby_duplicate_window([1, 2, 3, 1], 3))  # True
````

## Key Takeaways

| Pattern | When to Use |
|---------|-------------|
| **Hash map with index** | Track last occurrence |
| **Sliding window set** | Fixed-size window for duplicates |
