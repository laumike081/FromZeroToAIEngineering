## Problem Statement

Given two non-negative integers `low` and `high`. Return the count of odd numbers between `low` and `high` (inclusive).

### Examples

**Example 1:**
```
Input: low = 3, high = 7
Output: 3
Explanation: The odd numbers between 3 and 7 are [3,5,7].
```

**Example 2:**
```
Input: low = 8, high = 10
Output: 1
Explanation: The odd numbers between 8 and 10 are [9].
```

### Constraints

- `0 <= low <= high <= 10^9`

---

## Approach: Math Formula

### Intuition

Count of odd numbers from 0 to n = `(n + 1) // 2`

Count in range [low, high] = count(0 to high) - count(0 to low-1)

### Complexity
- **Time**: O(1)
- **Space**: O(1)

````python
def count_odds(low: int, high: int) -> int:
    """Math formula: O(1)"""
    # Count of odd numbers from 0 to n is (n + 1) // 2
    def count_to(n):
        return (n + 1) // 2
    
    return count_to(high) - count_to(low - 1)

# Test
print(count_odds(3, 7))   # 3
print(count_odds(8, 10))  # 1
````

---

## Simpler Formula

````python
def count_odds_simple(low: int, high: int) -> int:
    """Even simpler: half the range, +1 if either endpoint is odd"""
    return (high - low) // 2 + (1 if low % 2 == 1 or high % 2 == 1 else 0)

# Or even simpler:
def count_odds_v2(low: int, high: int) -> int:
    return (high + 1) // 2 - low // 2

print(count_odds_v2(3, 7))   # 3
print(count_odds_v2(8, 10))  # 1
````

## Walkthrough

**Example**: low=3, high=7

```
count_to(7) = (7+1)//2 = 4  (odds: 1,3,5,7)
count_to(2) = (2+1)//2 = 1  (odds: 1)
Result = 4 - 1 = 3
```

## Key Takeaways

| Pattern | When to Use |
|---------|-------------|
| **Prefix count subtraction** | Count in range = count(0,high) - count(0,low-1) |
| **Odd count formula** | `(n+1)//2` gives count of odds from 0 to n |
