# 3. Longest Substring Without Repeating Characters

**Difficulty**: Medium  
**Topics**: Hash Table, String, Sliding Window  
**Link**: [LeetCode 3](https://leetcode.com/problems/longest-substring-without-repeating-characters/)

---

## Problem Statement

Given a string `s`, find the length of the **longest substring** without repeating characters.

### Examples

**Example 1:**
```
Input: s = "abcabcbb"
Output: 3
Explanation: The answer is "abc", with the length of 3.
```

**Example 2:**
```
Input: s = "bbbbb"
Output: 1
Explanation: The answer is "b", with the length of 1.
```

**Example 3:**
```
Input: s = "pwwkew"
Output: 3
Explanation: The answer is "wke", with the length of 3.
```

### Constraints

- `0 <= s.length <= 5 * 10^4`
- `s` consists of English letters, digits, symbols and spaces.

---

## Approach 1: Sliding Window with Hash Map

### Intuition

Use a **sliding window** `[left, right]` that always contains unique characters.

- Expand `right` one character at a time
- When a duplicate is found, **jump `left`** to just past the previous occurrence of that character
- Track each character's **last seen index** in a hash map

**Key**: We jump `left` directly instead of sliding one step at a time — O(n) instead of O(n²).

### Complexity
- **Time**: O(n)
- **Space**: O(min(n, m)) where m = charset size

````python
def length_of_longest_substring(s: str) -> int:
    """Sliding window with hash map: O(n) time"""
    last_seen = {}  # char -> last index seen
    left = 0
    max_len = 0
    
    for right, char in enumerate(s):
        # If char was seen and is inside current window
        if char in last_seen and last_seen[char] >= left:
            left = last_seen[char] + 1  # Jump left past duplicate
        
        last_seen[char] = right
        max_len = max(max_len, right - left + 1)
    
    return max_len

# Test
print(f"'abcabcbb': {length_of_longest_substring('abcabcbb')}")  # 3
print(f"'bbbbb': {length_of_longest_substring('bbbbb')}")         # 1
print(f"'pwwkew': {length_of_longest_substring('pwwkew')}")       # 3
print(f"'': {length_of_longest_substring('')}")                   # 0
````

---

## Walkthrough

Input: `"abcabcbb"`

Each step: look up current char in `last_seen` → decide whether to move `left` → update `last_seen`

| right | char | left | Action | last_seen (full dict after step) | max_len |
|-------|------|------|--------|----------------------------------|---------|
| 0 | 'a' | 0 | first time, no move | {a:0} | 1 |
| 1 | 'b' | 0 | first time, no move | {a:0, b:1} | 2 |
| 2 | 'c' | 0 | first time, no move | {a:0, b:1, c:2} | 3 |
| 3 | 'a' | **1** | last_seen['a']=0 ≥ left=0 → left=0+1=**1** | {a:**3**, b:1, c:2} | 3 |
| 4 | 'b' | **2** | last_seen['b']=1 ≥ left=1 → left=1+1=**2** | {a:3, b:**4**, c:2} | 3 |
| 5 | 'c' | **3** | last_seen['c']=2 ≥ left=2 → left=2+1=**3** | {a:3, b:4, c:**5**} | 3 |
| 6 | 'b' | **5** | last_seen['b']=4 ≥ left=3 → left=4+1=**5** | {a:3, b:**6**, c:5} | 3 |
| 7 | 'b' | **7** | last_seen['b']=6 ≥ left=5 → left=6+1=**7** | {a:3, b:**7**, c:5} | 3 |

**Result**: 3

## Approach 2: Sliding Window with Set

Simpler logic — use a set to track characters in the current window, shrink from left when a duplicate is found.

````python
def length_of_longest_substring_set(s: str) -> int:
    """Sliding window with set: O(n) time, O(min(n,m)) space"""
    window = set()
    left = 0
    max_len = 0
    
    for right, char in enumerate(s):
        while char in window:
            window.remove(s[left])
            left += 1
        
        window.add(char)
        max_len = max(max_len, right - left + 1)
    
    return max_len

print(f"Set method 'abcabcbb': {length_of_longest_substring_set('abcabcbb')}")  # 3
print(f"Set method 'pwwkew': {length_of_longest_substring_set('pwwkew')}")       # 3
````

---

## Why `last_seen[char] >= left` Matters

Without this check, `left` could move **backwards**, breaking the window.

```
s = "abba"

right=3, char='a', last_seen['a']=0
  Current left = 2 (after seeing duplicate 'b')
  
  Without check: left = 0+1 = 1  ← WRONG! Moves left backward
  With check:    0 < left=2, so skip → left stays at 2  ✓
```

## Approach 3: Sliding Window with Character Array

### Intuition

Replace the dictionary with a **fixed-size integer array** indexed by ASCII value.

- `last_seen[ord(char)]` replaces `last_seen[char]`
- Array lookup is faster in practice than dict hashing — no hash computation overhead
- Size 128 covers all ASCII characters; use 256 for extended ASCII

### Complexity
- **Time**: O(n)
- **Space**: O(1) — fixed array of size 128, independent of input

````python
def length_of_longest_substring_array(s: str) -> int:
    """Character array instead of dict: O(n) time, O(1) space"""
    # Array of size 128 for all ASCII chars, initialized to -1 (not seen)
    last_seen = [-1] * 128
    left = 0
    max_len = 0

    for right, char in enumerate(s):
        ascii_val = ord(char)
        # If char was seen and is inside current window
        if last_seen[ascii_val] >= left:
            left = last_seen[ascii_val] + 1  # Jump left past duplicate

        last_seen[ascii_val] = right
        max_len = max(max_len, right - left + 1)

    return max_len

# Test
print(f"Array method 'abcabcbb': {length_of_longest_substring_array('abcabcbb')}")  # 3
print(f"Array method 'bbbbb':    {length_of_longest_substring_array('bbbbb')}")      # 1
print(f"Array method 'pwwkew':   {length_of_longest_substring_array('pwwkew')}")     # 3
print(f"Array method 'abba':     {length_of_longest_substring_array('abba')}")       # 2
````

---

## Comparison

| Approach | Time | Space | Pros | Cons |
|----------|------|-------|------|------|
| **Hash Map** | O(n) | O(m) | Clean, handles any char | Dict hashing overhead |
| **Set** | O(n) | O(m) | Simplest logic | Inner `while` loop |
| **Character Array** | O(n) | O(1) | Fastest — direct index, no hashing | Limited to ASCII (128/256) |

**Best choice**: **Character Array** for maximum performance; **Hash Map** for Unicode support.

## Key Takeaways

| Pattern | Key Point |
|---------|----------|
| **Sliding window** | Expand right, shrink left to maintain valid window |
| **Last seen index** | Jump `left` directly to skip duplicates in O(1) |
| **Guard condition** | Check `last_seen[char] >= left` to avoid backward movement |
| **Window length** | `right - left + 1` at each step |
