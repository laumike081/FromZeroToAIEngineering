## Problem Statement

We are playing the Guess Game. I pick a number from `1` to `n`. You have to guess which number I picked.

Every time you guess wrong, I will tell you whether the number I picked is higher or lower than your guess.

You call a pre-defined API `int guess(int num)`, which returns three possible results:
- `-1`: Your guess is higher than the number I picked.
- `1`: Your guess is lower than the number I picked.
- `0`: Your guess is equal to the number I picked.

Return the number that I picked.

### Constraints

- `1 <= n <= 2^31 - 1`

```python
# The guess API is already defined for you.
# def guess(num: int) -> int:

def guessNumber(n: int) -> int:
    """Binary search: O(log n)"""
    left, right = 1, n
    
    while left <= right:
        mid = (left + right) // 2
        result = guess(mid)
        
        if result == 0:
            return mid
        elif result == -1:
            right = mid - 1
        else:
            left = mid + 1
    
    return -1
```
