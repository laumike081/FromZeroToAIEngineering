## Problem Statement

There is a robot on an `m x n` grid. The robot is initially located at the top-left corner. The robot tries to move to the bottom-right corner. The robot can only move either down or right at any point in time.

Given the two integers `m` and `n`, return the number of possible unique paths that the robot can take to reach the bottom-right corner.

### Examples

**Example 1:**
```
Input: m = 3, n = 7
Output: 28
```

### Constraints

- `1 <= m, n <= 100`

---

## Approach 1: Dynamic Programming

### Intuition

`dp[i][j]` = number of ways to reach cell (i, j) = `dp[i-1][j] + dp[i][j-1]`

### Complexity
- **Time**: O(m * n)
- **Space**: O(n) with space optimization

````python
def unique_paths(m: int, n: int) -> int:
    """DP with O(n) space"""
    dp = [1] * n
    
    for _ in range(1, m):
        for j in range(1, n):
            dp[j] += dp[j-1]
    
    return dp[n-1]

# Test
print(unique_paths(3, 7))  # 28
print(unique_paths(3, 2))  # 3
````

---

## Approach 2: Math (Combinatorics)

### Intuition

Total moves = (m-1) down + (n-1) right = C(m+n-2, m-1)

````python
from math import comb

def unique_paths_math(m: int, n: int) -> int:
    """Combinatorics: O(min(m,n))"""
    return comb(m + n - 2, m - 1)

print(unique_paths_math(3, 7))  # 28
````
