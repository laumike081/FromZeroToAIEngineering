# 206. Reverse Linked List

## Problem Statement

Given the `head` of a singly linked list, reverse the list, and return *the reversed list*.

### Examples

**Example 1:**
```
Input: head = [1,2,3,4,5]
Output: [5,4,3,2,1]
```

**Example 2:**
```
Input: head = [1,2]
Output: [2,1]
```

**Example 3:**
```
Input: head = []
Output: []
```

### Constraints

- The number of nodes in the list is in the range `[0, 5000]`
- `-5000 <= Node.val <= 5000`

````python
# Definition for singly-linked list
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
    
    def __repr__(self):
        result = []
        node = self
        while node:
            result.append(str(node.val))
            node = node.next
        return " -> ".join(result)

def create_list(values):
    """Helper to create linked list from array"""
    if not values:
        return None
    head = ListNode(values[0])
    current = head
    for val in values[1:]:
        current.next = ListNode(val)
        current = current.next
    return head
````

## Approach 1: Iterative (Three Pointers)

### Intuition

Use three pointers to reverse links one by one:
- `prev`: previously processed node (starts as None)
- `curr`: current node being processed
- `next`: next node to process (saved before reversing)

### Algorithm

```
1 -> 2 -> 3 -> 4 -> 5 -> None

Step 1: None <- 1    2 -> 3 -> 4 -> 5
Step 2: None <- 1 <- 2    3 -> 4 -> 5
Step 3: None <- 1 <- 2 <- 3    4 -> 5
Step 4: None <- 1 <- 2 <- 3 <- 4    5
Step 5: None <- 1 <- 2 <- 3 <- 4 <- 5
```

### Complexity
- **Time**: O(n)
- **Space**: O(1)

````python
def reverse_list_iterative(head: ListNode) -> ListNode:
    """Iterative: O(n) time, O(1) space"""
    prev = None
    curr = head
    
    while curr:
        next_temp = curr.next  # Save next node
        curr.next = prev       # Reverse the link
        prev = curr            # Move prev forward
        curr = next_temp       # Move curr forward
    
    return prev  # prev is now the new head

# Test
head = create_list([1, 2, 3, 4, 5])
print(f"Original: {head}")
reversed_head = reverse_list_iterative(head)
print(f"Reversed: {reversed_head}")
````

## Approach 2: Recursive

### Intuition

1. Recursively reverse the rest of the list
2. Make the next node point back to current node
3. Set current node's next to None

### Complexity
- **Time**: O(n)
- **Space**: O(n) - recursion stack

````python
def reverse_list_recursive(head: ListNode) -> ListNode:
    """Recursive: O(n) time, O(n) space"""
    # Base case: empty list or single node
    if not head or not head.next:
        return head
    
    # Recursively reverse the rest
    new_head = reverse_list_recursive(head.next)
    
    # Make next node point back to current
    head.next.next = head
    head.next = None
    
    return new_head

# Test
head = create_list([1, 2, 3, 4, 5])
print(f"Original: {head}")
reversed_head = reverse_list_recursive(head)
print(f"Reversed: {reversed_head}")
````

## Walkthrough (Iterative)

```
Initial: 1 -> 2 -> 3 -> None
         prev=None, curr=1

Step 1:
  next_temp = 2
  1.next = None (was 2)
  prev = 1
  curr = 2
  
  State: None <- 1    2 -> 3 -> None

Step 2:
  next_temp = 3
  2.next = 1 (was 3)
  prev = 2
  curr = 3
  
  State: None <- 1 <- 2    3 -> None

Step 3:
  next_temp = None
  3.next = 2 (was None)
  prev = 3
  curr = None
  
  State: None <- 1 <- 2 <- 3

Return prev (3) as new head
```

````python
def reverse_list_verbose(head: ListNode) -> ListNode:
    """Verbose version showing each step"""
    prev = None
    curr = head
    step = 0
    
    while curr:
        print(f"Step {step}: curr={curr.val}, prev={prev.val if prev else None}")
        
        next_temp = curr.next
        curr.next = prev
        prev = curr
        curr = next_temp
        
        step += 1
    
    print(f"Done: new head = {prev.val}")
    return prev

head = create_list([1, 2, 3])
reverse_list_verbose(head)
````

---

## Common Mistakes

1. **Losing reference to next node** - must save `curr.next` before overwriting
2. **Returning wrong node** - return `prev`, not `curr` (which is None)
3. **Not handling empty list** - check `if not head`

## Key Takeaways

| Pattern | When to Use |
|---------|-------------|
| **Three pointers (prev, curr, next)** | In-place linked list reversal |
| **Save before overwrite** | Always save `next` before changing links |
| **Iterative vs Recursive** | Iterative for O(1) space, recursive for cleaner code |
