## Problem Statement

Two strings are considered **close** if you can attain one from the other using the following operations:

- **Operation 1**: Swap any two existing characters.
- **Operation 2**: Transform every occurrence of one existing character into another existing character, and do the same with the other character.

You can use the operations on either string as many times as necessary.

Given two strings, `word1` and `word2`, return `true` if `word1` and `word2` are **close**, and `false` otherwise.

### Examples

**Example 1:**
```
Input: word1 = "abc", word2 = "bca"
Output: true
Explanation: Swap 'a' and 'b', then swap 'b' and 'c'.
```

**Example 2:**
```
Input: word1 = "a", word2 = "aa"
Output: false
```

**Example 3:**
```
Input: word1 = "cabbba", word2 = "abbccc"
Output: true
Explanation: Transform 'c' to 'b' and 'b' to 'c'.
```

### Constraints

- `1 <= word1.length, word2.length <= 10^5`
- `word1` and `word2` contain only lowercase English letters

---

## Approach: Compare Character Sets and Frequency Distributions

### Intuition

Two strings are close if:
1. They have the **same set of characters** (can't create new characters)
2. They have the **same frequency distribution** (sorted frequencies match)

### Complexity
- **Time**: O(n)
- **Space**: O(1) - at most 26 characters

````python
from collections import Counter

def close_strings(word1: str, word2: str) -> bool:
    """Compare sets and sorted frequencies"""
    if len(word1) != len(word2):
        return False
    
    count1 = Counter(word1)
    count2 = Counter(word2)
    
    # Same characters
    if set(count1.keys()) != set(count2.keys()):
        return False
    
    # Same frequency distribution
    return sorted(count1.values()) == sorted(count2.values())

# Test
print(close_strings("abc", "bca"))       # True
print(close_strings("a", "aa"))          # False
print(close_strings("cabbba", "abbccc")) # True
````

---

## Key Takeaways

| Pattern | When to Use |
|---------|-------------|
| **Same character set** | Characters can be swapped but not created |
| **Same frequency distribution** | Frequencies can be reassigned between chars |
