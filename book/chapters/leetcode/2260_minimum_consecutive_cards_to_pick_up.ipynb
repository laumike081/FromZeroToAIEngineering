## Problem Statement

You are given an integer array `cards` where `cards[i]` represents the value of the i-th card. A pair of cards are **matching** if the two cards have the same value.

Return the **minimum number of consecutive cards** you have to pick up to contain a pair of matching cards. If it is impossible to have matching cards, return `-1`.

### Examples

**Example 1:**
```
Input: cards = [3,4,2,3,4,7]
Output: 4
Explanation: We can pick up the cards [3,4,2,3] which contain a matching pair of cards with value 3.
Note that picking up [4,2,3,4] is also optimal.
```

**Example 2:**
```
Input: cards = [1,0,5,3]
Output: -1
Explanation: There is no way to pick up a set of consecutive cards that contain a pair of matching cards.
```

### Constraints

- `1 <= cards.length <= 10^5`
- `0 <= cards[i] <= 10^6`

---

## Approach: Hash Map (Last Seen Index)

### Intuition

For each card value, we want the **shortest window** between any two occurrences of that value.

**Algorithm**:
1. Track the last index where each card value was seen
2. When we see a duplicate, compute the window length: `i - last_seen[card] + 1`
3. Track the minimum across all duplicates

### Complexity
- **Time**: O(n)
- **Space**: O(n)

````python
def minimum_card_pickup(cards: list[int]) -> int:
    """Hash map tracking last seen index: O(n) time and space"""
    last_seen = {}  # card_value -> last index seen
    min_len = float('inf')
    
    for i, card in enumerate(cards):
        if card in last_seen:
            # Window from last occurrence to current (inclusive)
            min_len = min(min_len, i - last_seen[card] + 1)
        last_seen[card] = i  # Always update to latest index
    
    return min_len if min_len != float('inf') else -1

# Test
print(f"[3,4,2,3,4,7]: {minimum_card_pickup([3,4,2,3,4,7])}")  # 4
print(f"[1,0,5,3]: {minimum_card_pickup([1,0,5,3])}")           # -1
print(f"[0,0]: {minimum_card_pickup([0,0])}")                   # 2
````

---

## Walkthrough

Input: `[3, 4, 2, 3, 4, 7]`

| Index | Card | In Map? | Window Length | Min Len | last_seen update |
|-------|------|---------|---------------|---------|------------------|
| 0 | 3 | No | - | ∞ | {3: 0} |
| 1 | 4 | No | - | ∞ | {3:0, 4:1} |
| 2 | 2 | No | - | ∞ | {3:0, 4:1, 2:2} |
| 3 | 3 | Yes (0) | 3-0+1 = **4** | 4 | {3:3, 4:1, 2:2} |
| 4 | 4 | Yes (1) | 4-1+1 = **4** | 4 | {3:3, 4:4, 2:2} |
| 5 | 7 | No | - | 4 | {3:3, 4:4, 2:2, 7:5} |

**Result**: 4

> **Key**: We always update `last_seen` to the **latest** index, so consecutive duplicates use the nearest pair.

---

## Why Update last_seen Every Time?

Consider: `[3, 3, 3]`

- At index 1: window = 1-0+1 = 2 → min = 2, update last_seen[3] = 1
- At index 2: window = 2-1+1 = 2 → min = 2

If we **didn't** update:
- At index 2: window = 2-0+1 = 3 → miss the optimal answer!

Always updating ensures we find the **closest** pair.

---

## Alternative: Using defaultdict

````python
from collections import defaultdict

def minimum_card_pickup_v2(cards: list[int]) -> int:
    """Store all indices per card, then find min gap."""
    indices = defaultdict(list)
    
    for i, card in enumerate(cards):
        indices[card].append(i)
    
    min_len = float('inf')
    for idx_list in indices.values():
        for j in range(1, len(idx_list)):
            min_len = min(min_len, idx_list[j] - idx_list[j-1] + 1)
    
    return min_len if min_len != float('inf') else -1

print(f"v2: {minimum_card_pickup_v2([3,4,2,3,4,7])}")  # 4
````

## Comparison

| Approach | Time | Space | Notes |
|----------|------|-------|-------|
| **Last Seen (Approach 1)** | O(n) | O(n) | Single pass, optimal |
| **All Indices (Approach 2)** | O(n) | O(n) | Two passes, more code |

Approach 1 is preferred — simpler and single pass.

## Key Takeaways

| Pattern | Key Point |
|---------|----------|
| **Last seen index** | Always update to latest — finds closest duplicate |
| **Window length** | `i - last_seen[val] + 1` (inclusive on both ends) |
| **No match** | Return -1 when `min_len` stays at infinity |
