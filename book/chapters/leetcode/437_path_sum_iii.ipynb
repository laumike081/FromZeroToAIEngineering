## Problem Statement

Given the `root` of a binary tree and an integer `targetSum`, return the number of paths where the sum of the values along the path equals `targetSum`.

The path does not need to start or end at the root or a leaf, but it must go downwards.

### Examples

**Example 1:**
```
Input: root = [10,5,-3,3,2,null,11,3,-2,null,1], targetSum = 8
Output: 3
```

### Constraints

- The number of nodes is in the range `[0, 1000]`
- `-10^9 <= Node.val <= 10^9`
- `-1000 <= targetSum <= 1000`

````python
from collections import defaultdict

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def path_sum(root: TreeNode, targetSum: int) -> int:
    """Prefix sum with DFS: O(n)"""
    prefix_count = defaultdict(int)
    prefix_count[0] = 1
    
    def dfs(node, curr_sum):
        if not node:
            return 0
        
        curr_sum += node.val
        # Count paths ending at current node
        count = prefix_count[curr_sum - targetSum]
        
        # Add current sum to map
        prefix_count[curr_sum] += 1
        
        # Recurse
        count += dfs(node.left, curr_sum)
        count += dfs(node.right, curr_sum)
        
        # Backtrack
        prefix_count[curr_sum] -= 1
        
        return count
    
    return dfs(root, 0)
````

## Key Takeaways

| Pattern | When to Use |
|---------|-------------|
| **Prefix sum + hash map** | Count subarrays/paths with target sum |
| **Backtracking** | Remove state when leaving subtree |
