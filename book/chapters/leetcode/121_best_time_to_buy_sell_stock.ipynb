## Problem Statement

You are given an array `prices` where `prices[i]` is the price of a given stock on the `i`th day.

You want to maximize your profit by choosing a **single day** to buy one stock and choosing a **different day in the future** to sell that stock.

Return the *maximum profit* you can achieve from this transaction. If you cannot achieve any profit, return `0`.

### Examples

**Example 1:**
```
Input: prices = [7,1,5,3,6,4]
Output: 5
Explanation: Buy on day 2 (price = 1) and sell on day 5 (price = 6), profit = 6-1 = 5.
```

**Example 2:**
```
Input: prices = [7,6,4,3,1]
Output: 0
Explanation: No profitable transaction possible.
```

### Constraints

- `1 <= prices.length <= 10^5`
- `0 <= prices[i] <= 10^4`

---

## Approach: Track Minimum Price

### Intuition

For each day, the maximum profit if we sell on that day is:
```
profit = current_price - minimum_price_so_far
```

We track:
1. **Minimum price seen so far** (best day to buy)
2. **Maximum profit** (best profit if we sell today)

### Complexity
- **Time**: O(n) - single pass
- **Space**: O(1) - two variables

````python
def max_profit(prices: list[int]) -> int:
    """Track min price: O(n) time, O(1) space"""
    min_price = float('inf')
    max_profit = 0
    
    for price in prices:
        # Update minimum price seen so far
        min_price = min(min_price, price)
        
        # Calculate profit if we sell today
        profit = price - min_price
        
        # Update maximum profit
        max_profit = max(max_profit, profit)
    
    return max_profit

# Test
print(max_profit([7, 1, 5, 3, 6, 4]))  # 5
print(max_profit([7, 6, 4, 3, 1]))     # 0
````

````python
def max_profit_verbose(prices: list[int]) -> int:
    """Verbose version to show the process"""
    min_price = float('inf')
    max_profit = 0
    
    print(f"{'Day':<5} {'Price':<7} {'Min':<7} {'Profit':<8} {'MaxProfit':<10}")
    print("-" * 40)
    
    for i, price in enumerate(prices):
        min_price = min(min_price, price)
        profit = price - min_price
        max_profit = max(max_profit, profit)
        
        print(f"{i:<5} {price:<7} {min_price:<7} {profit:<8} {max_profit:<10}")
    
    return max_profit

max_profit_verbose([7, 1, 5, 3, 6, 4])
````

## Alternative: Kadane's Algorithm Perspective

This problem can also be viewed as finding the **maximum subarray sum** of daily price changes.

````python
def max_profit_kadane(prices: list[int]) -> int:
    """Kadane's algorithm on price differences"""
    max_profit = 0
    current_profit = 0
    
    for i in range(1, len(prices)):
        # Daily change
        change = prices[i] - prices[i - 1]
        
        # Either continue the current transaction or start fresh
        current_profit = max(0, current_profit + change)
        
        max_profit = max(max_profit, current_profit)
    
    return max_profit

# Test
print(max_profit_kadane([7, 1, 5, 3, 6, 4]))  # 5
print(max_profit_kadane([7, 6, 4, 3, 1]))     # 0
````

## Key Takeaways

| Pattern | When to Use |
|---------|-------------|
| **Track running min/max** | Optimization over array with ordering constraint |
| **Single pass** | When future decisions depend only on past minimum |
| **Kadane's variant** | Maximum subarray problems |
