## Problem Statement

Given a string `text`, you want to use the characters of `text` to form as many instances of the word **"balloon"** as possible.

You can use each character in `text` **at most once**. Return the maximum number of instances that can be formed.

### Examples

**Example 1:**
```
Input: text = "nlaebolko"
Output: 1
```

**Example 2:**
```
Input: text = "loonbalxballpoon"
Output: 2
```

**Example 3:**
```
Input: text = "leetcode"
Output: 0
```

### Constraints

- `1 <= text.length <= 10^4`
- `text` consists of lower case English letters only.

---

## Approach: Character Counting

### Intuition

The word "balloon" requires:
- `b`: 1
- `a`: 1
- `l`: 2
- `o`: 2
- `n`: 1

Count characters in `text`, then find how many complete "balloon" words we can form.

The limiting factor is the character with the smallest ratio of (available / required).

### Complexity
- **Time**: O(n)
- **Space**: O(1) - only 26 letters

````python
from collections import Counter

def max_number_of_balloons(text: str) -> int:
    """Character counting: O(n) time, O(1) space"""
    count = Counter(text)
    
    # "balloon" needs: b=1, a=1, l=2, o=2, n=1
    return min(
        count['b'],
        count['a'],
        count['l'] // 2,  # Need 2 l's per balloon
        count['o'] // 2,  # Need 2 o's per balloon
        count['n']
    )

# Test
print(f"'nlaebolko': {max_number_of_balloons('nlaebolko')}")
print(f"'loonbalxballpoon': {max_number_of_balloons('loonbalxballpoon')}")
print(f"'leetcode': {max_number_of_balloons('leetcode')}")
````

## Approach 2: Array Counting (No Hash Map)

Using a fixed-size array instead of Counter for slightly better performance.

````python
def max_number_of_balloons_array(text: str) -> int:
    """Array counting without hash map: O(n) time, O(1) space"""
    # Count only the characters we need: b, a, l, o, n
    count = [0] * 5  # [b, a, l, o, n]
    
    for char in text:
        if char == 'b':
            count[0] += 1
        elif char == 'a':
            count[1] += 1
        elif char == 'l':
            count[2] += 1
        elif char == 'o':
            count[3] += 1
        elif char == 'n':
            count[4] += 1
    
    # "balloon" needs: b=1, a=1, l=2, o=2, n=1
    return min(
        count[0],      # b
        count[1],      # a
        count[2] // 2, # l (need 2)
        count[3] // 2, # o (need 2)
        count[4]       # n
    )

# Test
print(f"Array method 'nlaebolko': {max_number_of_balloons_array('nlaebolko')}")
print(f"Array method 'loonbalxballpoon': {max_number_of_balloons_array('loonbalxballpoon')}")
````

## Approach 3: Single Pass with Dictionary

A cleaner single-pass approach using a dictionary with predefined keys.

````python
def max_number_of_balloons_dict(text: str) -> int:
    """Single pass with dictionary: O(n) time, O(1) space"""
    # Only track characters in "balloon"
    count = {'b': 0, 'a': 0, 'l': 0, 'o': 0, 'n': 0}
    
    for char in text:
        if char in count:
            count[char] += 1
    
    # Divide l and o by 2 since we need 2 of each
    count['l'] //= 2
    count['o'] //= 2
    
    return min(count.values())

# Test
print(f"Dict method: {max_number_of_balloons_dict('loonbalxballpoon')}")
````

---

## Comparison of Approaches

| Approach | Time | Space | Pros | Cons |
|----------|------|-------|------|------|
| **Counter** | O(n) | O(26) | Clean, Pythonic | Counts all 26 chars |
| **Array** | O(n) | O(5) | Minimal space | Less readable |
| **Dict with keys** | O(n) | O(5) | Clean + efficient | Slightly more code |
| **Generic** | O(n) | O(k) | Works for any pattern | Overkill for this problem |

**Best choice**: The **Dict with keys** approach is the best balance of readability and efficiency - it only tracks the 5 characters we need and uses `min(count.values())` for a clean solution.

````python
def max_instances(text: str, pattern: str) -> int:
    """Generic version for any pattern word."""
    text_count = Counter(text)
    pattern_count = Counter(pattern)
    
    result = float('inf')
    for char, required in pattern_count.items():
        available = text_count.get(char, 0)
        result = min(result, available // required)
    
    return result if result != float('inf') else 0

print(f"Generic 'balloon': {max_instances('loonbalxballpoon', 'balloon')}")
print(f"Generic 'hello': {max_instances('helloworld', 'hello')}")
````
