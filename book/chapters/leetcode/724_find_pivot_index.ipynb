## Problem Statement

Given an array of integers `nums`, calculate the **pivot index** of this array.

The **pivot index** is the index where the sum of all the numbers **strictly** to the left of the index is equal to the sum of all the numbers **strictly** to the index's right.

If the index is on the left edge of the array, then the left sum is `0`. The same applies to the right edge.

Return the **leftmost pivot index**. If no such index exists, return `-1`.

### Examples

**Example 1:**
```
Input: nums = [1,7,3,6,5,6]
Output: 3
Explanation: Left sum = 1+7+3 = 11, Right sum = 5+6 = 11
```

**Example 2:**
```
Input: nums = [1,2,3]
Output: -1
```

### Constraints

- `1 <= nums.length <= 10^4`
- `-1000 <= nums[i] <= 1000`

---

## Approach: Prefix Sum

### Intuition

- Total sum = left_sum + nums[i] + right_sum
- At pivot: left_sum == right_sum
- So: left_sum == (total - nums[i]) / 2

### Complexity
- **Time**: O(n)
- **Space**: O(1)

````python
def pivot_index(nums: list[int]) -> int:
    """Prefix sum: O(n) time, O(1) space"""
    total = sum(nums)
    left_sum = 0
    
    for i, num in enumerate(nums):
        # right_sum = total - left_sum - num
        if left_sum == total - left_sum - num:
            return i
        left_sum += num
    
    return -1

# Test
print(pivot_index([1, 7, 3, 6, 5, 6]))  # 3
print(pivot_index([1, 2, 3]))           # -1
print(pivot_index([2, 1, -1]))          # 0
````

---

## Key Takeaways

| Pattern | When to Use |
|---------|-------------|
| **Prefix sum** | Compare left/right sums efficiently |
| **right = total - left - current** | Avoid computing right sum separately |
