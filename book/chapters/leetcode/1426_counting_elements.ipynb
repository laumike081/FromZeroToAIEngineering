## Problem Statement

Given an integer array `arr`, count how many elements `x` there are, such that `x + 1` is also in `arr`. If there are duplicates in `arr`, count them separately.

### Examples

**Example 1:**
```
Input: arr = [1,2,3]
Output: 2
Explanation: 1 and 2 are counted cause 2 and 3 are in arr.
```

**Example 2:**
```
Input: arr = [1,1,3,3,5,5,7,7]
Output: 0
Explanation: No numbers are counted, cause there is no 2, 4, 6, or 8 in arr.
```

**Example 3:**
```
Input: arr = [1,1,2,2]
Output: 2
Explanation: Two 1s are counted cause 2 is in arr.
```

### Constraints

- `1 <= arr.length <= 1000`
- `0 <= arr[i] <= 1000`

---

## Approach: Hash Set

### Intuition

1. Create a set of all elements for O(1) lookup
2. For each element, check if `x + 1` exists in the set
3. Count duplicates separately

### Complexity
- **Time**: O(n)
- **Space**: O(n)

````python
def count_elements(arr: list[int]) -> int:
    """Hash set: O(n) time and space"""
    num_set = set(arr)
    count = 0
    
    for x in arr:
        if x + 1 in num_set:
            count += 1
    
    return count

# Test
print(count_elements([1, 2, 3]))           # 2
print(count_elements([1, 1, 3, 3, 5, 5]))  # 0
print(count_elements([1, 1, 2, 2]))        # 2
````

---

## Walkthrough

Input: `[1, 1, 2, 2]`

Set: `{1, 2}`

| x | x + 1 | In set? | Count |
|---|-------|---------|-------|
| 1 | 2 | Yes | 1 |
| 1 | 2 | Yes | 2 |
| 2 | 3 | No | 2 |
| 2 | 3 | No | 2 |

**Result**: 2

---

## One-Liner

````python
def count_elements_oneliner(arr: list[int]) -> int:
    s = set(arr)
    return sum(1 for x in arr if x + 1 in s)

print(count_elements_oneliner([1, 2, 3]))  # 2
````
