## Problem Statement

You are given a **0-indexed** array `nums` consisting of positive integers. You can choose two indices `i` and `j` such that `i != j`, and the **sum of digits** of the number `nums[i]` is equal to that of `nums[j]`.

Return the **maximum** value of `nums[i] + nums[j]` that you can obtain over all possible indices `i` and `j` that satisfy the conditions. If no such pair exists, return `-1`.

### Examples

**Example 1:**
```
Input: nums = [18,43,36,13,7]
Output: 54
Explanation:
The pairs (i, j) that satisfy the conditions are:
- (0, 2), both nums[0]=18 and nums[2]=36 have a digit sum of 9. Their sum is 18 + 36 = 54.
- (1, 4), both nums[1]=43 and nums[4]=7 have a digit sum of 7. Their sum is 43 + 7 = 50.
So the maximum sum is 54.
```

**Example 2:**
```
Input: nums = [10,12,19,14]
Output: -1
Explanation: There are no two numbers with the same digit sum.
```

### Constraints

- `1 <= nums.length <= 10^5`
- `1 <= nums[i] <= 10^9`

---

## Approach: Hash Map (Best Per Digit Sum)

### Intuition

For each digit sum group, we only need the **two largest** numbers to maximize their sum.

**Algorithm**:
1. Compute digit sum for each number
2. For each digit sum, track the largest number seen so far in a hash map
3. When a second number with the same digit sum is found, compute their sum and update the answer
4. Always keep the larger of the two in the hash map

### Complexity
- **Time**: O(n log m) where m is the max number (log m digits per number)
- **Space**: O(n)

````python
def digit_sum(n: int) -> int:
    """Compute sum of digits of n."""
    total = 0
    while n:
        total += n % 10
        n //= 10
    return total

def maximum_sum(nums: list[int]) -> int:
    """Hash map tracking best value per digit sum: O(n log m)"""
    best = {}  # digit_sum -> largest number with that digit sum
    result = -1
    
    for num in nums:
        ds = digit_sum(num)
        if ds in best:
            result = max(result, best[ds] + num)
            best[ds] = max(best[ds], num)  # Keep the larger
        else:
            best[ds] = num
    
    return result

# Test
print(f"[18,43,36,13,7]: {maximum_sum([18,43,36,13,7])}")  # 54
print(f"[10,12,19,14]: {maximum_sum([10,12,19,14])}")       # -1
print(f"[1,2,3,4]: {maximum_sum([1,2,3,4])}")               # 7 (3+4)
````

## Walkthrough

Input: `[18, 43, 36, 13, 7]`

| Num | Digit Sum | In Map? | Candidate Sum | Result | best map |
|-----|-----------|---------|---------------|--------|----------|
| 18 | 9 | No | - | -1 | {9: 18} |
| 43 | 7 | No | - | -1 | {9:18, 7:43} |
| 36 | 9 | Yes (18) | 18+36 = **54** | 54 | {9:36, 7:43} |
| 13 | 4 | No | - | 54 | {9:36, 7:43, 4:13} |
| 7 | 7 | Yes (43) | 43+7 = 50 | 54 | {9:36, 7:43, 4:13} |

**Result**: 54

## Alternative: Group All Values, Then Find Top 2

````python
from collections import defaultdict

def maximum_sum_v2(nums: list[int]) -> int:
    """Group by digit sum, then find top 2 per group."""
    groups = defaultdict(list)
    
    for num in nums:
        groups[digit_sum(num)].append(num)
    
    result = -1
    for group in groups.values():
        if len(group) >= 2:
            group.sort(reverse=True)
            result = max(result, group[0] + group[1])
    
    return result

print(f"v2: {maximum_sum_v2([18,43,36,13,7])}")  # 54
````

## Helper: Python One-liner for Digit Sum

````python
# Pythonic digit sum using string conversion
digit_sum_str = lambda n: sum(int(d) for d in str(n))

print(f"digit_sum(18) = {digit_sum_str(18)}")   # 9
print(f"digit_sum(43) = {digit_sum_str(43)}")   # 7
print(f"digit_sum(999) = {digit_sum_str(999)}") # 27
````

## Comparison

| Approach | Time | Space | Notes |
|----------|------|-------|-------|
| **Best per group (Approach 1)** | O(n log m) | O(n) | Single pass, optimal |
| **Group all, sort (Approach 2)** | O(n log n) | O(n) | Simpler but slower |

Approach 1 is preferred — no sorting needed, single pass.
