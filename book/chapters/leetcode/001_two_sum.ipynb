# 1. Two Sum

**Difficulty**: Easy  
**Topics**: Array, Hash Table  
**Link**: [LeetCode 1](https://leetcode.com/problems/two-sum/)

---

## Problem Statement

Given an array of integers `nums` and an integer `target`, return *indices of the two numbers such that they add up to `target`*.

You may assume that each input would have **exactly one solution**, and you may not use the same element twice.

You can return the answer in any order.

### Examples

**Example 1:**
```
Input: nums = [2,7,11,15], target = 9
Output: [0,1]
Explanation: nums[0] + nums[1] = 2 + 7 = 9
```

**Example 2:**
```
Input: nums = [3,2,4], target = 6
Output: [1,2]
```

**Example 3:**
```
Input: nums = [3,3], target = 6
Output: [0,1]
```

### Constraints

- `2 <= nums.length <= 10^4`
- `-10^9 <= nums[i] <= 10^9`
- `-10^9 <= target <= 10^9`
- Only one valid answer exists.

---

## Approach 1: Brute Force

### Intuition
Check every pair of numbers to see if they sum to target.

### Complexity
- **Time**: O(n²) - nested loops
- **Space**: O(1) - no extra space

````python
def two_sum_brute(nums: list[int], target: int) -> list[int]:
    """Brute force: O(n²) time, O(1) space"""
    n = len(nums)
    for i in range(n):
        for j in range(i + 1, n):
            if nums[i] + nums[j] == target:
                return [i, j]
    return []  # No solution found

# Test
print(two_sum_brute([2, 7, 11, 15], 9))  # [0, 1]
print(two_sum_brute([3, 2, 4], 6))       # [1, 2]
print(two_sum_brute([3, 3], 6))          # [0, 1]
````

---

## Approach 2: Hash Map (Optimal)

### Intuition

For each number `num`, we need to find if `target - num` exists in the array.

Instead of searching with a nested loop (O(n)), use a hash map for O(1) lookup.

**Key insight**: As we iterate, store each number's index. For the current number, check if its complement (`target - num`) was seen before.

### Algorithm

1. Create an empty hash map: `{value: index}`
2. For each `(index, num)` in array:
   - Calculate `complement = target - num`
   - If `complement` in hash map, return `[hash_map[complement], index]`
   - Otherwise, add `num: index` to hash map

### Complexity
- **Time**: O(n) - single pass
- **Space**: O(n) - hash map storage

````python
def two_sum(nums: list[int], target: int) -> list[int]:
    """Hash map: O(n) time, O(n) space"""
    seen = {}  # value -> index
    
    for i, num in enumerate(nums):
        complement = target - num
        
        if complement in seen:
            return [seen[complement], i]
        
        seen[num] = i
    
    return []  # No solution found

# Test
print(two_sum([2, 7, 11, 15], 9))  # [0, 1]
print(two_sum([3, 2, 4], 6))       # [1, 2]
print(two_sum([3, 3], 6))          # [0, 1]
````

## Walkthrough

Let's trace through `nums = [2, 7, 11, 15]`, `target = 9`:

| Step | num | complement | seen | Action |
|------|-----|------------|------|--------|
| 0 | 2 | 7 | {} | 7 not in seen, add {2: 0} |
| 1 | 7 | 2 | {2: 0} | **2 in seen!** Return [0, 1] |

Found in just 2 iterations!

````python
def two_sum_verbose(nums: list[int], target: int) -> list[int]:
    """Verbose version to show the process"""
    seen = {}
    
    for i, num in enumerate(nums):
        complement = target - num
        print(f"Step {i}: num={num}, complement={complement}, seen={seen}")
        
        if complement in seen:
            print(f"  → Found! complement {complement} at index {seen[complement]}")
            return [seen[complement], i]
        
        seen[num] = i
        print(f"  → Added {num}: {i} to seen")
    
    return []

two_sum_verbose([2, 7, 11, 15], 9)
````

---

## Common Mistakes

1. **Using the same element twice**: `[3, 2, 4]` with target `6` - can't use index 0 twice
2. **Returning values instead of indices**: Problem asks for indices, not the numbers
3. **Not handling duplicates**: `[3, 3]` with target `6` should return `[0, 1]`

## Key Takeaways

| Pattern | When to Use |
|---------|-------------|
| **Hash Map for complement** | Finding pairs that sum to target |
| **Single pass** | Build hash map while searching |
| **Trade space for time** | O(n) space to achieve O(n) time |
