# 2390. Removing Stars From a String

**Difficulty**: Medium

## Problem Statement

You are given a string `s`, which contains stars `*`.

In one operation, you can:
- Choose a star in `s`.
- Remove the closest **non-star** character to its **left**, as well as remove the star itself.

Return the string after **all** stars have been removed.

**Note:**
- The input will be generated such that the operation is always possible.
- It can be shown that the resulting string will always be unique.

### Examples

**Example 1:**
```
Input: s = "leet**cod*e"
Output: "lecoe"
Explanation: 
- Remove 't' and first '*': "lee*cod*e"
- Remove 'e' and second '*': "lecod*e"
- Remove 'd' and third '*': "lecoe"
```

**Example 2:**
```
Input: s = "erase*****"
Output: ""
Explanation: The entire string is removed.
```

### Constraints

- `1 <= s.length <= 10^5`
- `s` consists of lowercase English letters and stars `*`
- The operation above can be performed on `s`

---

## Approach: Stack

### Intuition

Think of `*` as a backspace:
- Letters: push to stack
- Star: pop from stack (remove previous character)

### Complexity
- **Time**: O(n)
- **Space**: O(n)

````python
def remove_stars(s: str) -> str:
    """Stack approach: O(n) time and space"""
    stack = []
    
    for char in s:
        if char == '*':
            stack.pop()  # Remove previous character
        else:
            stack.append(char)
    
    return ''.join(stack)

# Test
print(remove_stars("leet**cod*e"))  # "lecoe"
print(remove_stars("erase*****"))   # ""
````

---

## Walkthrough

Input: `"leet**cod*e"`

| char | Action | Stack |
|------|--------|-------|
| 'l' | push | ['l'] |
| 'e' | push | ['l', 'e'] |
| 'e' | push | ['l', 'e', 'e'] |
| 't' | push | ['l', 'e', 'e', 't'] |
| '*' | pop 't' | ['l', 'e', 'e'] |
| '*' | pop 'e' | ['l', 'e'] |
| 'c' | push | ['l', 'e', 'c'] |
| 'o' | push | ['l', 'e', 'c', 'o'] |
| 'd' | push | ['l', 'e', 'c', 'o', 'd'] |
| '*' | pop 'd' | ['l', 'e', 'c', 'o'] |
| 'e' | push | ['l', 'e', 'c', 'o', 'e'] |

**Result**: `"lecoe"`

````python
def remove_stars_verbose(s: str) -> str:
    """Verbose version"""
    stack = []
    
    print(f"Input: '{s}'\n")
    
    for i, char in enumerate(s):
        if char == '*':
            removed = stack.pop()
            print(f"Step {i}: '*' → pop '{removed}'")
        else:
            stack.append(char)
            print(f"Step {i}: '{char}' → push")
        print(f"         Stack: {''.join(stack)}")
    
    return ''.join(stack)

remove_stars_verbose("leet**cod*e")
````

## Alternative: Two Pointers (In-Place)

For languages with mutable strings, we can do this in-place.

````python
def remove_stars_two_pointers(s: str) -> str:
    """Two pointers using list"""
    result = list(s)
    write = 0  # Write pointer
    
    for char in s:
        if char == '*':
            write -= 1  # Move back (like backspace)
        else:
            result[write] = char
            write += 1
    
    return ''.join(result[:write])

# Test
print(remove_stars_two_pointers("leet**cod*e"))  # "lecoe"
````

---

## Common Mistakes

1. **Popping from empty stack** - problem guarantees valid input
2. **Using string concatenation** - O(n^2), use list instead
3. **Processing right to left** - stars remove character to the LEFT
