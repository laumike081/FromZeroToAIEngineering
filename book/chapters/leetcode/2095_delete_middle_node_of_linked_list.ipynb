# 2095. Delete the Middle Node of a Linked List

**Difficulty**: Medium  
**Topics**: Linked List, Two Pointers  
**Link**: [LeetCode 2095](https://leetcode.com/problems/delete-the-middle-node-of-a-linked-list/)

## Problem Statement

You are given the `head` of a linked list. Delete the **middle node**, and return the `head` of the modified linked list.

The **middle node** of a linked list of size `n` is the `⌊n / 2⌋`th node from the start using 0-based indexing.

### Examples

**Example 1:**
```
Input: head = [1,3,4,7,1,2,6]
Output: [1,3,4,1,2,6]
Explanation: Node 7 (index 3) is the middle.
```

**Example 2:**
```
Input: head = [1,2,3,4]
Output: [1,2,4]
Explanation: Node 3 (index 2) is the middle.
```

**Example 3:**
```
Input: head = [2,1]
Output: [2]
Explanation: Node 1 (index 1) is the middle.
```

### Constraints

- The number of nodes is in the range `[1, 10^5]`
- `1 <= Node.val <= 10^5`

---

## Approach 1: Slow and Fast Pointers

### Intuition

Use the classic slow/fast pointer technique:
- **Fast pointer** moves 2 steps
- **Slow pointer** moves 1 step
- Keep track of `prev` to delete the middle node

When fast reaches the end, slow will be at the middle position.

### Why it works

For a list of length `n`:
- Fast moves `2k` steps while slow moves `k` steps
- When fast can't move 2 steps anymore, slow is at `⌊n/2⌋`

### Complexity
- **Time**: O(n) - single traversal
- **Space**: O(1) - constant extra space

````python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def delete_middle(head: ListNode) -> ListNode:
    """Slow/Fast pointers: O(n) time, O(1) space"""
    if not head or not head.next:
        return None
    
    prev = None
    slow = head
    fast = head
    
    while fast and fast.next:
        prev = slow
        slow = slow.next
        fast = fast.next.next
    
    # slow is at middle, prev is before middle
    prev.next = slow.next
    return head

# --- Test ---
# Create list: 1 -> 3 -> 4 -> 7 -> 1 -> 2 -> 6
h = ListNode(1, ListNode(3, ListNode(4, ListNode(7, ListNode(1, ListNode(2, ListNode(6)))))))
new_head = delete_middle(h)
result = []
while new_head:
    result.append(new_head.val)
    new_head = new_head.next
print(f"After deletion: {result}")  # [1, 3, 4, 1, 2, 6]
````

---

## Walkthrough

Input: `[1,3,4,7,1,2,6]`

| Step | prev | slow | fast | Action | List after step |
|------|------|------|------|--------|----------------|
| 0 | None | 1 | 1 | Initialize | 1→3→4→7→1→2→6 |
| 1 | 1 | 3 | 4 | Move: prev=slow, slow=slow.next, fast=fast.next.next | 1→3→4→7→1→2→6 |
| 2 | 3 | 4 | 7 | Move: prev=slow, slow=slow.next, fast=fast.next.next | 1→3→4→7→1→2→6 |
| 3 | 4 | 7 | 2 | Move: prev=slow, slow=slow.next, fast=fast.next.next | 1→3→4→7→1→2→6 |
| 4 | 7 | 1 | 6 | Move: prev=slow, slow=slow.next, fast=fast.next.next | 1→3→4→7→1→2→6 |
| 5 | 1 | 2 | None | fast=None, stop. slow is at middle (index 3) | 1→3→4→7→1→2→6 |
| 6 | 1 | 2 | None | Delete: prev.next = slow.next (skip node 7) | 1→3→4→1→2→6 |

**Result**: `[1,3,4,1,2,6]` - middle node `7` deleted.

**Key observation**: When fast reaches the end, slow is exactly at `⌊n/2⌋` position.

## Approach 2: Count and Delete

### Intuition

Two-pass approach:
1. First pass: count total nodes
2. Second pass: navigate to middle-1 and delete middle

### Complexity
- **Time**: O(n) - two traversals
- **Space**: O(1) - constant extra space

````python
def delete_middle_count(head: ListNode) -> ListNode:
    """Count then delete: O(n) time, O(1) space"""
    if not head or not head.next:
        return None
    
    # Count nodes
    count = 0
    curr = head
    while curr:
        count += 1
        curr = curr.next
    
    # Find middle index
    middle_idx = count // 2
    
    # Navigate to middle-1
    prev = None
    curr = head
    for _ in range(middle_idx):
        prev = curr
        curr = curr.next
    
    # Delete middle
    prev.next = curr.next
    return head

# --- Test ---
h = ListNode(1, ListNode(3, ListNode(4, ListNode(7, ListNode(1, ListNode(2, ListNode(6)))))))
new_head = delete_middle_count(h)
result = []
while new_head:
    result.append(new_head.val)
    new_head = new_head.next
print(f"After deletion: {result}")  # [1, 3, 4, 1, 2, 6]
````

## Approach 3: Optimized Fast Pointer

### Intuition

Start fast at `head.next.next` to skip ahead, so slow ends up at node before middle.

### Why it works

By initializing fast two steps ahead:
- When fast reaches end, slow is at `middle-1`
- Can directly delete `slow.next` without tracking prev

### Complexity
- **Time**: O(n) - single traversal
- **Space**: O(1) - constant extra space

````python
def delete_middle_optimized(head: ListNode) -> ListNode:
    """Optimized fast pointer: O(n) time, O(1) space"""
    if not head or not head.next:
        return None
    
    slow = head
    fast = head.next.next
    
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
    
    # slow is at middle-1, delete slow.next
    slow.next = slow.next.next
    return head

# --- Test ---
h = ListNode(1, ListNode(3, ListNode(4, ListNode(7, ListNode(1, ListNode(2, ListNode(6)))))))
new_head = delete_middle_optimized(h)
result = []
while new_head:
    result.append(new_head.val)
    new_head = new_head.next
print(f"After deletion: {result}")  # [1, 3, 4, 1, 2, 6]
````

---

## Comparison

| Approach | Time | Space | Pros | Cons |
|----------|------|-------|------|------|
| **Slow/Fast + prev** | O(n) | O(1) | Intuitive, clear logic | Need extra `prev` pointer |
| **Count then delete** | O(n) | O(1) | Simple to understand | Two traversals |
| **Optimized fast** | O(n) | O(1) | No `prev` needed, clean | Less intuitive initialization |

**Best choice**: **Optimized fast pointer** for cleanest code.

## Key Takeaways

| Pattern | When to Use |
|---------|-------------|
| **Slow/Fast pointers** | Find middle, detect cycles |
| **Track previous** | Delete node by skipping |
| **Optimized initialization** | Start fast ahead to position slow |
| **Count then navigate** | When two passes are acceptable |

**Remember**: For middle deletion, you need the node **before** the middle to skip it.
