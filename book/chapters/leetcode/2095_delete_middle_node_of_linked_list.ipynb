# 2095. Delete the Middle Node of a Linked List

**Difficulty**: Medium  
**Topics**: Linked List, Two Pointers  
**Link**: [LeetCode 2095](https://leetcode.com/problems/delete-the-middle-node-of-a-linked-list/)

---

## Problem Statement

You are given the `head` of a linked list. Delete the **middle node**, and return the `head` of the modified linked list.

The **middle node** of a linked list of size `n` is the `⌊n / 2⌋`th node from the start using 0-based indexing.

### Examples

**Example 1:**
```
Input: head = [1,3,4,7,1,2,6]
Output: [1,3,4,1,2,6]
Explanation: Node 7 (index 3) is the middle.
```

### Constraints

- The number of nodes is in the range `[1, 10^5]`
- `1 <= Node.val <= 10^5`

---

## Approach: Slow and Fast Pointers

### Intuition

Use slow/fast pointers. When fast reaches end, slow is at middle. Keep track of previous node to delete.

### Complexity
- **Time**: O(n)
- **Space**: O(1)

````python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def delete_middle(head: ListNode) -> ListNode:
    """Slow/Fast pointers: O(n)"""
    if not head or not head.next:
        return None
    
    slow = head
    fast = head.next.next
    
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
    
    # slow.next is the middle node
    slow.next = slow.next.next
    return head
````

---

## Key Takeaways

| Pattern | When to Use |
|---------|-------------|
| **Slow/Fast pointers** | Find middle of linked list |
| **Track previous** | Delete node by skipping |
