# 1133. Largest Unique Number

## Problem Statement

Given an integer array `nums`, return the largest integer that only occurs once. If no integer occurs once, return `-1`.

### Examples

**Example 1:**
```
Input: nums = [5,7,3,9,4,9,8,3,1]
Output: 8
Explanation: The maximum integer in the array is 9 but it is repeated.
The number 8 occurs only once, so it is the answer.
```

**Example 2:**
```
Input: nums = [9,9,8,8]
Output: -1
Explanation: There is no number that occurs only once.
```

### Constraints

- `1 <= nums.length <= 2000`
- `0 <= nums[i] <= 1000`

---

## Approach: Hash Map Counting

### Intuition

1. Count the frequency of each number using a hash map
2. Find the maximum number that has a count of exactly 1
3. Return -1 if no unique number exists

### Complexity
- **Time**: O(n)
- **Space**: O(n)

````python
from collections import Counter

def largest_unique_number(nums: list[int]) -> int:
    """Hash map counting: O(n) time and space"""
    count = Counter(nums)
    
    result = -1
    for num, freq in count.items():
        if freq == 1:
            result = max(result, num)
    
    return result

# Test
print(f"[5,7,3,9,4,9,8,3,1]: {largest_unique_number([5,7,3,9,4,9,8,3,1])}")
print(f"[9,9,8,8]: {largest_unique_number([9,9,8,8])}")
````

## One-Liner

````python
def largest_unique_number_oneliner(nums: list[int]) -> int:
    count = Counter(nums)
    unique = [n for n, c in count.items() if c == 1]
    return max(unique) if unique else -1

print(f"One-liner: {largest_unique_number_oneliner([5,7,3,9,4,9,8,3,1])}")
````

## Alternative: Sorting Approach

````python
def largest_unique_number_sort(nums: list[int]) -> int:
    """Sort and scan: O(n log n) time, O(1) extra space"""
    nums.sort(reverse=True)
    n = len(nums)
    
    i = 0
    while i < n:
        # Count occurrences of current number
        j = i
        while j < n and nums[j] == nums[i]:
            j += 1
        
        # If only one occurrence, it's our answer
        if j - i == 1:
            return nums[i]
        
        i = j
    
    return -1

print(f"Sorting: {largest_unique_number_sort([5,7,3,9,4,9,8,3,1])}")
````
