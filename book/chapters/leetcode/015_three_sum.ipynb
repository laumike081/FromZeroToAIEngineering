# 15. 3Sum

**Difficulty**: Medium  
**Topics**: Array, Two Pointers, Sorting  
**Link**: [LeetCode 15](https://leetcode.com/problems/3sum/)

---

## Problem Statement

Given an integer array `nums`, return all the triplets `[nums[i], nums[j], nums[k]]` such that `i != j`, `i != k`, and `j != k`, and `nums[i] + nums[j] + nums[k] == 0`.

Notice that the solution set must not contain duplicate triplets.

### Examples

**Example 1:**
```
Input: nums = [-1,0,1,2,-1,-4]
Output: [[-1,-1,2],[-1,0,1]]
Explanation: 
nums[0] + nums[1] + nums[2] = (-1) + 0 + 1 = 0.
nums[1] + nums[2] + nums[4] = 0 + 1 + (-1) = 0.
nums[0] + nums[3] + nums[4] = (-1) + 2 + (-1) = 0.
The distinct triplets are [-1,0,1] and [-1,-1,2].
```

**Example 2:**
```
Input: nums = [0,1,1]
Output: []
Explanation: The only possible triplet does not sum up to 0.
```

**Example 3:**
```
Input: nums = [0,0,0]
Output: [[0,0,0]]
```

### Constraints

- `3 <= nums.length <= 3000`
- `-10^5 <= nums[i] <= 10^5`

---

## Approach: Sort + Two Pointers

### Intuition

1. **Sort the array** - enables two-pointer technique and easy duplicate skipping
2. **Fix one element** - iterate through array, fix `nums[i]`
3. **Two Sum on remainder** - find two numbers that sum to `-nums[i]`

### Key Insight

After sorting, use two pointers (left, right) to find pairs:
- If sum < 0: move left pointer right (need larger values)
- If sum > 0: move right pointer left (need smaller values)
- If sum == 0: found a triplet!

### Handling Duplicates

- Skip duplicate values for the fixed element `i`
- Skip duplicate values when moving left/right pointers

### Complexity
- **Time**: O(n²) - O(n log n) sort + O(n²) two-pointer search
- **Space**: O(1) or O(n) depending on sort implementation

````python
def three_sum(nums: list[int]) -> list[list[int]]:
    """Sort + Two Pointers: O(n²) time"""
    nums.sort()
    result = []
    n = len(nums)
    
    for i in range(n - 2):
        # Skip duplicates for i
        if i > 0 and nums[i] == nums[i - 1]:
            continue
        
        # Early termination: if smallest is positive, no solution
        if nums[i] > 0:
            break
        
        # Two pointers for remaining elements
        left, right = i + 1, n - 1
        target = -nums[i]
        
        while left < right:
            current_sum = nums[left] + nums[right]
            
            if current_sum < target:
                left += 1
            elif current_sum > target:
                right -= 1
            else:
                # Found a triplet
                result.append([nums[i], nums[left], nums[right]])
                
                # Skip duplicates for left and right
                while left < right and nums[left] == nums[left + 1]:
                    left += 1
                while left < right and nums[right] == nums[right - 1]:
                    right -= 1
                
                left += 1
                right -= 1
    
    return result

# Test
print(three_sum([-1, 0, 1, 2, -1, -4]))  # [[-1,-1,2], [-1,0,1]]
print(three_sum([0, 1, 1]))              # []
print(three_sum([0, 0, 0]))              # [[0,0,0]]
````

## Walkthrough

Input: `[-1, 0, 1, 2, -1, -4]`

**Step 1: Sort**
```
[-4, -1, -1, 0, 1, 2]
```

**Step 2: Iterate with two pointers**

| i | nums[i] | target | left | right | sum | action |
|---|---------|--------|------|-------|-----|--------|
| 0 | -4 | 4 | 1 | 5 | -1+2=1 | 1<4, left++ |
| 0 | -4 | 4 | 2 | 5 | -1+2=1 | 1<4, left++ |
| 0 | -4 | 4 | 3 | 5 | 0+2=2 | 2<4, left++ |
| 0 | -4 | 4 | 4 | 5 | 1+2=3 | 3<4, left++ |
| 0 | -4 | 4 | 5 | 5 | done | |
| 1 | -1 | 1 | 2 | 5 | -1+2=1 | **Found [-1,-1,2]** |
| 1 | -1 | 1 | 3 | 4 | 0+1=1 | **Found [-1,0,1]** |
| 2 | -1 | skip | | | | duplicate of i=1 |
| 3 | 0 | 0 | 4 | 5 | 1+2=3 | 3>0, right-- |
| 3 | 0 | 0 | 4 | 4 | done | |

**Result**: `[[-1,-1,2], [-1,0,1]]`

````python
def three_sum_verbose(nums: list[int]) -> list[list[int]]:
    """Verbose version to show the process"""
    nums.sort()
    print(f"Sorted: {nums}")
    result = []
    n = len(nums)
    
    for i in range(n - 2):
        if i > 0 and nums[i] == nums[i - 1]:
            print(f"\ni={i}: Skip duplicate {nums[i]}")
            continue
        
        if nums[i] > 0:
            print(f"\ni={i}: {nums[i]} > 0, break early")
            break
        
        print(f"\ni={i}: Fix nums[{i}]={nums[i]}, target={-nums[i]}")
        left, right = i + 1, n - 1
        target = -nums[i]
        
        while left < right:
            current_sum = nums[left] + nums[right]
            print(f"  left={left}({nums[left]}), right={right}({nums[right]}), sum={current_sum}")
            
            if current_sum < target:
                left += 1
            elif current_sum > target:
                right -= 1
            else:
                triplet = [nums[i], nums[left], nums[right]]
                print(f"  → Found triplet: {triplet}")
                result.append(triplet)
                
                while left < right and nums[left] == nums[left + 1]:
                    left += 1
                while left < right and nums[right] == nums[right - 1]:
                    right -= 1
                left += 1
                right -= 1
    
    return result

three_sum_verbose([-1, 0, 1, 2, -1, -4])
````

## Common Mistakes

1. **Not handling duplicates** - results in duplicate triplets
2. **Wrong duplicate skip logic** - must skip AFTER finding, not before
3. **Off-by-one errors** - careful with `i > 0` check
4. **Forgetting to move both pointers** after finding a triplet

## Key Takeaways

| Pattern | When to Use |
|---------|-------------|
| **Sort + Two Pointers** | Finding pairs/triplets with sum constraint |
| **Fix one, search two** | Reduce 3Sum to 2Sum |
| **Skip duplicates after sort** | Avoid duplicate results |
| **Early termination** | If smallest > target, no solution possible |
