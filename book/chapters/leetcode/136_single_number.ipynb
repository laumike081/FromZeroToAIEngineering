# 136. Single Number

**Difficulty**: Easy  
**Topics**: Array, Bit Manipulation  
**Link**: [LeetCode 136](https://leetcode.com/problems/single-number/)

---

## Problem Statement

Given a **non-empty** array of integers `nums`, every element appears twice except for one. Find that single one.

You must implement a solution with a linear runtime complexity and use only constant extra space.

### Examples

**Example 1:**
```
Input: nums = [2,2,1]
Output: 1
```

### Constraints

- `1 <= nums.length <= 3 * 10^4`
- `-3 * 10^4 <= nums[i] <= 3 * 10^4`
- Each element appears twice except for one element which appears once

---

## Approach: XOR

### Intuition

XOR properties:
- `a ^ a = 0`
- `a ^ 0 = a`
- XOR is commutative and associative

XOR all numbers together - pairs cancel out, leaving the single number.

### Complexity
- **Time**: O(n)
- **Space**: O(1)

````python
def single_number(nums: list[int]) -> int:
    """XOR: O(n) time, O(1) space"""
    result = 0
    for num in nums:
        result ^= num
    return result

# Test
print(single_number([2, 2, 1]))        # 1
print(single_number([4, 1, 2, 1, 2]))  # 4
````
