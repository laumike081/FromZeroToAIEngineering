## Problem Statement

Given a **0-indexed** `n x n` integer matrix `grid`, return the number of pairs `(ri, cj)` such that row `ri` and column `cj` are equal.

A row and column pair is considered equal if they contain the same elements in the same order.

### Examples

**Example 1:**
```
Input: grid = [[3,2,1],[1,7,6],[2,7,7]]
Output: 1
Explanation: Row 2 = [2,7,7], Column 1 = [2,7,7]
```

**Example 2:**
```
Input: grid = [[3,1,2,2],[1,4,4,5],[2,4,2,2],[2,4,2,2]]
Output: 3
```

### Constraints

- `n == grid.length == grid[i].length`
- `1 <= n <= 200`
- `1 <= grid[i][j] <= 10^5`

---

## Approach: Hash Map with Tuple Keys

### Intuition

1. Store each row as a tuple in a hash map with count
2. For each column, convert to tuple and check if it exists in map

### Complexity
- **Time**: O(n^2)
- **Space**: O(n^2)

````python
from collections import Counter

def equal_pairs(grid: list[list[int]]) -> int:
    """Hash map approach: O(n^2)"""
    n = len(grid)
    
    # Count rows as tuples
    row_count = Counter(tuple(row) for row in grid)
    
    # Count matching columns
    count = 0
    for c in range(n):
        col = tuple(grid[r][c] for r in range(n))
        count += row_count[col]
    
    return count

# Test
print(equal_pairs([[3,2,1],[1,7,6],[2,7,7]]))  # 1
print(equal_pairs([[3,1,2,2],[1,4,4,5],[2,4,2,2],[2,4,2,2]]))  # 3
````

## Key Takeaways

| Pattern | When to Use |
|---------|-------------|
| **Tuple as hash key** | Compare sequences in hash map |
| **Counter for duplicates** | Handle multiple matching rows |
