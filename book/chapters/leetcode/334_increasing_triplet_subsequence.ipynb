## Problem Statement

Given an integer array `nums`, return `true` if there exists a triple of indices `(i, j, k)` such that `i < j < k` and `nums[i] < nums[j] < nums[k]`. If no such indices exists, return `false`.

### Examples

**Example 1:**
```
Input: nums = [1,2,3,4,5]
Output: true
Explanation: Any triplet where i < j < k is valid.
```

**Example 2:**
```
Input: nums = [5,4,3,2,1]
Output: false
Explanation: No triplet exists.
```

**Example 3:**
```
Input: nums = [2,1,5,0,4,6]
Output: true
Explanation: The triplet (3, 4, 5) is valid: nums[3]=0 < nums[4]=4 < nums[5]=6.
```

### Constraints

- `1 <= nums.length <= 5 * 10^5`
- `-2^31 <= nums[i] <= 2^31 - 1`

**Follow up:** Could you implement a solution that runs in O(n) time and O(1) space?

---

## Approach: Track Two Smallest Values

### Intuition

Track the smallest and second smallest values seen so far:
- `first`: smallest value
- `second`: smallest value greater than `first`

If we find a value greater than `second`, we have a valid triplet.

### Complexity
- **Time**: O(n)
- **Space**: O(1)

````python
def increasing_triplet(nums: list[int]) -> bool:
    """Track two smallest: O(n) time, O(1) space"""
    first = second = float('inf')
    
    for num in nums:
        if num <= first:
            first = num  # Smallest so far
        elif num <= second:
            second = num  # Second smallest (and > first)
        else:
            # num > second > first, found triplet!
            return True
    
    return False

# Test
print(increasing_triplet([1, 2, 3, 4, 5]))  # True
print(increasing_triplet([5, 4, 3, 2, 1]))  # False
print(increasing_triplet([2, 1, 5, 0, 4, 6]))  # True
````

## Walkthrough

Input: `[2, 1, 5, 0, 4, 6]`

| num | first | second | Action |
|-----|-------|--------|--------|
| 2 | 2 | inf | Update first |
| 1 | 1 | inf | Update first |
| 5 | 1 | 5 | Update second |
| 0 | 0 | 5 | Update first |
| 4 | 0 | 4 | Update second |
| 6 | - | - | **6 > 4 > (some previous first)** → True |

**Key insight**: Even though `first` changed to 0, `second=5` still represents a valid "second element" because there was a smaller element before it.

````python
def increasing_triplet_verbose(nums: list[int]) -> bool:
    """Verbose version"""
    first = second = float('inf')
    
    print(f"Input: {nums}\n")
    
    for i, num in enumerate(nums):
        if num <= first:
            first = num
            print(f"i={i}, num={num}: Update first={first}")
        elif num <= second:
            second = num
            print(f"i={i}, num={num}: Update second={second}")
        else:
            print(f"i={i}, num={num}: Found triplet! {num} > {second}")
            return True
    
    return False

increasing_triplet_verbose([2, 1, 5, 0, 4, 6])
````

---

## Common Mistakes

1. **Thinking first must come before second in array** - `second` implicitly has a smaller element before it
2. **Using `<` instead of `<=`** - we want strictly increasing, so equal values don't count
