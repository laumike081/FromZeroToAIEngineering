## Problem Statement

You are given two 0-indexed integer arrays `nums1` and `nums2` of equal length `n` and a positive integer `k`. You must choose a subsequence of indices from `nums1` of length `k`.

For chosen indices `i0, i1, ..., ik-1`, your score is defined as:
- The sum of the selected elements from `nums1` multiplied with the minimum of the selected elements from `nums2`.

Return the **maximum** possible score.

### Constraints

- `n == nums1.length == nums2.length`
- `1 <= n <= 10^5`
- `0 <= nums1[i], nums2[i] <= 10^5`

```python
import heapq

def max_score(nums1: list[int], nums2: list[int], k: int) -> int:
    """Sort by nums2 desc, min heap for nums1: O(n log n)"""
    # Pair and sort by nums2 descending
    pairs = sorted(zip(nums2, nums1), reverse=True)
    
    heap = []  # Min heap for nums1 values
    sum1 = 0
    max_score = 0
    
    for num2, num1 in pairs:
        heapq.heappush(heap, num1)
        sum1 += num1
        
        if len(heap) > k:
            sum1 -= heapq.heappop(heap)
        
        if len(heap) == k:
            max_score = max(max_score, sum1 * num2)
    
    return max_score

# Test
print(max_score([1,3,3,2], [2,1,3,4], 3))  # 12
```
