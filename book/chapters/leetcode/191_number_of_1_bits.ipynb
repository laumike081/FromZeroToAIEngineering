## Problem Statement

Write a function that takes the binary representation of a positive integer and returns the number of set bits it has (also known as the **Hamming weight**).

### Examples

**Example 1:**
```
Input: n = 11
Output: 3
Explanation: The input binary string 1011 has three '1' bits.
```

**Example 2:**
```
Input: n = 128
Output: 1
Explanation: The input binary string 10000000 has one '1' bit.
```

**Example 3:**
```
Input: n = 2147483645
Output: 30
```

### Constraints

- `1 <= n <= 2^31 - 1`

---

## Approach 1: Brian Kernighan's Algorithm

### Intuition

Key trick: `n & (n - 1)` removes the rightmost set bit.

Count how many times we can do this until `n` becomes 0.

### Complexity
- **Time**: O(k) where k = number of set bits
- **Space**: O(1)

````python
def hamming_weight(n: int) -> int:
    """Brian Kernighan's Algorithm: O(k) where k = set bits"""
    count = 0
    
    while n:
        n &= (n - 1)  # Remove rightmost set bit
        count += 1
    
    return count

# Test
print(hamming_weight(11))   # 3 (binary: 1011)
print(hamming_weight(128))  # 1 (binary: 10000000)
print(hamming_weight(7))    # 3 (binary: 111)
````

---

## Walkthrough

Input: `n = 11` (binary: `1011`)

| Step | n (binary) | n - 1 | n & (n-1) | count |
|------|------------|-------|-----------|-------|
| 0 | 1011 | 1010 | 1010 | 1 |
| 1 | 1010 | 1001 | 1000 | 2 |
| 2 | 1000 | 0111 | 0000 | 3 |
| 3 | 0000 | - | done | **3** |

````python
def hamming_weight_verbose(n: int) -> int:
    """Verbose version"""
    count = 0
    print(f"Input: {n} = {bin(n)}\n")
    
    while n:
        prev = n
        n &= (n - 1)
        count += 1
        print(f"Step {count}: {prev:04b} & {prev-1:04b} = {n:04b}")
    
    print(f"\nTotal 1-bits: {count}")
    return count

hamming_weight_verbose(11)
````

## Approach 2: Check Each Bit

````python
def hamming_weight_loop(n: int) -> int:
    """Check each bit: O(32) time"""
    count = 0
    
    while n:
        count += n & 1  # Add last bit
        n >>= 1         # Shift right
    
    return count

print(hamming_weight_loop(11))  # 3
````

---

## Python Built-in

````python
def hamming_weight_builtin(n: int) -> int:
    """Using Python built-in"""
    return bin(n).count('1')

# Or even simpler in Python 3.10+
# return n.bit_count()

print(hamming_weight_builtin(11))  # 3
````

---

## Why `n & (n-1)` Works

```
n     = ...1000  (rightmost 1 followed by zeros)
n-1   = ...0111  (that 1 becomes 0, zeros become 1s)
n&n-1 = ...0000  (rightmost 1 is removed)
```
