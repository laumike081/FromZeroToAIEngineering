# 1493. Longest Subarray of 1's After Deleting One Element

**Difficulty**: Medium  
**Topics**: Array, Sliding Window, Dynamic Programming

---

## Problem Statement

Given a binary array `nums`, you should delete one element from it.

Return the size of the longest non-empty subarray containing only `1`'s in the resulting array. Return `0` if there is no such subarray.

### Examples

**Example 1:**
```
Input: nums = [1,1,0,1]
Output: 3
Explanation: Delete the 0 to get [1,1,1].
```

**Example 2:**
```
Input: nums = [0,1,1,1,0,1,1,0,1]
Output: 5
```

**Example 3:**
```
Input: nums = [1,1,1]
Output: 2
Explanation: Must delete one element.
```

### Constraints

- `1 <= nums.length <= 10^5`
- `nums[i]` is either `0` or `1`

---

## Approach: Sliding Window (Allow One Zero)

### Intuition

Find the longest window with at most one 0 (which we'll delete).

### Complexity
- **Time**: O(n)
- **Space**: O(1)

````python
def longest_subarray(nums: list[int]) -> int:
    """Sliding window with at most one zero: O(n)"""
    left = 0
    zeros = 0
    max_len = 0
    
    for right in range(len(nums)):
        if nums[right] == 0:
            zeros += 1
        
        # Shrink window if more than one zero
        while zeros > 1:
            if nums[left] == 0:
                zeros -= 1
            left += 1
        
        # Window size - 1 (must delete one element)
        max_len = max(max_len, right - left)
    
    return max_len

# Test
print(longest_subarray([1, 1, 0, 1]))              # 3
print(longest_subarray([0, 1, 1, 1, 0, 1, 1, 0, 1])) # 5
print(longest_subarray([1, 1, 1]))                 # 2
````

---

## Key Takeaways

| Pattern | When to Use |
|---------|-------------|
| **Sliding window with k allowed** | Max subarray with at most k violations |
| **Window size - 1** | Must delete one element |
