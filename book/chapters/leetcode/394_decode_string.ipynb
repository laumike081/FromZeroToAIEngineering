## Problem Statement

Given an encoded string, return its decoded string.

The encoding rule is: `k[encoded_string]`, where the `encoded_string` inside the square brackets is being repeated exactly `k` times. Note that `k` is guaranteed to be a positive integer.

You may assume that the input string is always valid; there are no extra white spaces, square brackets are well-formed, etc.

Furthermore, you may assume that the original data does not contain any digits and that digits are only for those repeat numbers, `k`.

### Examples

**Example 1:**
```
Input: s = "3[a]2[bc]"
Output: "aaabcbc"
```

**Example 2:**
```
Input: s = "3[a2[c]]"
Output: "accaccacc"
```

**Example 3:**
```
Input: s = "2[abc]3[cd]ef"
Output: "abcabccdcdcdef"
```

### Constraints

- `1 <= s.length <= 30`
- `s` consists of lowercase English letters, digits, and square brackets `'[]'`
- `s` is guaranteed to be a valid input
- All integers in `s` are in the range `[1, 300]`

## Approach 1: Stack

### Intuition

Use a stack to handle nested brackets:
1. When we see `[`, push current string and number onto stack
2. When we see `]`, pop and build: `prev_string + num * current_string`
3. Digits build the repeat count
4. Letters append to current string

### Complexity
- **Time**: O(n * max_k) where max_k is the maximum repeat count
- **Space**: O(n) for the stack

````python
def decode_string(s: str) -> str:
    """Stack approach: O(n) time and space"""
    stack = []  # stores (previous_string, repeat_count)
    current_string = ""
    current_num = 0
    
    for char in s:
        if char.isdigit():
            # Build number (could be multi-digit like "12")
            current_num = current_num * 10 + int(char)
        
        elif char == '[':
            # Save current state and start fresh
            stack.append((current_string, current_num))
            current_string = ""
            current_num = 0
        
        elif char == ']':
            # Pop and build result
            prev_string, num = stack.pop()
            current_string = prev_string + num * current_string
        
        else:
            # Regular letter
            current_string += char
    
    return current_string

# Test
print(decode_string("3[a]2[bc]"))      # "aaabcbc"
print(decode_string("3[a2[c]]"))       # "accaccacc"
print(decode_string("2[abc]3[cd]ef"))  # "abcabccdcdcdef"
````

## Walkthrough

Input: `"3[a2[c]]"`

| char | current_num | current_string | stack | Action |
|------|-------------|----------------|-------|--------|
| '3' | 3 | "" | [] | Build number |
| '[' | 0 | "" | [("", 3)] | Push, reset |
| 'a' | 0 | "a" | [("", 3)] | Append letter |
| '2' | 2 | "a" | [("", 3)] | Build number |
| '[' | 0 | "" | [("", 3), ("a", 2)] | Push, reset |
| 'c' | 0 | "c" | [("", 3), ("a", 2)] | Append letter |
| ']' | 0 | "acc" | [("", 3)] | Pop: "a" + 2*"c" |
| ']' | 0 | "accaccacc" | [] | Pop: "" + 3*"acc" |

**Result**: `"accaccacc"`

````python
def decode_string_verbose(s: str) -> str:
    """Verbose version showing the process"""
    stack = []
    current_string = ""
    current_num = 0
    
    print(f"Input: '{s}'\n")
    
    for i, char in enumerate(s):
        print(f"Step {i}: char='{char}'")
        
        if char.isdigit():
            current_num = current_num * 10 + int(char)
            print(f"  → Building number: {current_num}")
        
        elif char == '[':
            stack.append((current_string, current_num))
            print(f"  → Push ('{current_string}', {current_num}) to stack")
            current_string = ""
            current_num = 0
        
        elif char == ']':
            prev_string, num = stack.pop()
            current_string = prev_string + num * current_string
            print(f"  → Pop: '{prev_string}' + {num} * '{current_string}' = '{current_string}'")
        
        else:
            current_string += char
            print(f"  → Append letter: '{current_string}'")
        
        print(f"  Stack: {stack}, Current: '{current_string}'")
    
    return current_string

decode_string_verbose("3[a2[c]]")
````

## Approach 2: Recursion

### Intuition

Recursively decode nested brackets. Use an index pointer to track position.

````python
def decode_string_recursive(s: str) -> str:
    """Recursive approach"""
    def decode(index: int) -> tuple[str, int]:
        result = ""
        num = 0
        
        while index < len(s):
            char = s[index]
            
            if char.isdigit():
                num = num * 10 + int(char)
            
            elif char == '[':
                # Recurse to decode inner content
                decoded, index = decode(index + 1)
                result += num * decoded
                num = 0
            
            elif char == ']':
                # Return to caller
                return result, index
            
            else:
                result += char
            
            index += 1
        
        return result, index
    
    return decode(0)[0]

# Test
print(decode_string_recursive("3[a]2[bc]"))      # "aaabcbc"
print(decode_string_recursive("3[a2[c]]"))       # "accaccacc"
print(decode_string_recursive("2[abc]3[cd]ef"))  # "abcabccdcdcdef"
````

---

## Common Mistakes

1. **Multi-digit numbers** - "12[a]" means repeat 12 times, not 1 then 2
2. **Nested brackets** - must handle "3[a2[c]]" correctly
3. **Letters outside brackets** - "2[a]bc" should give "aabc"
4. **Forgetting to reset** - reset num and string after `[`

## Key Takeaways

| Pattern | When to Use |
|---------|-------------|
| **Stack for nested structures** | Matching brackets, parentheses |
| **Push on open, pop on close** | Save/restore state at nesting levels |
| **Build number digit by digit** | `num = num * 10 + digit` |
