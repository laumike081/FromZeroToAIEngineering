## Problem Statement

You are given a **0-indexed** array `nums` of `n` integers, and an integer `k`.

The **k-radius average** for a subarray of `nums` **centered** at some index `i` with the **radius** `k` is the average of **all** elements in `nums` between the indices `i - k` and `i + k` (**inclusive**).

Build and return an array `avgs` of length `n` where `avgs[i]` is the **k-radius average** for the subarray centered at index `i`.

If there are less than `k` elements before or after the index `i`, then the **k-radius average** is `-1`.

### Examples

**Example 1:**
```
Input: nums = [7,4,3,9,1,8,5,2,6], k = 3
Output: [-1,-1,-1,5,4,4,-1,-1,-1]
```

### Constraints

- `n == nums.length`
- `1 <= n <= 10^5`
- `0 <= nums[i], k <= 10^5`

---

## Approach: Sliding Window

### Complexity
- **Time**: O(n)
- **Space**: O(1) extra

````python
def get_averages(nums: list[int], k: int) -> list[int]:
    """Sliding window: O(n)"""
    n = len(nums)
    result = [-1] * n
    window_size = 2 * k + 1
    
    if window_size > n:
        return result
    
    # Initial window sum
    window_sum = sum(nums[:window_size])
    result[k] = window_sum // window_size
    
    # Slide window
    for i in range(k + 1, n - k):
        window_sum += nums[i + k] - nums[i - k - 1]
        result[i] = window_sum // window_size
    
    return result

# Test
print(get_averages([7,4,3,9,1,8,5,2,6], 3))  # [-1,-1,-1,5,4,4,-1,-1,-1]
````

---

## Key Takeaways

| Pattern | When to Use |
|---------|-------------|
| **Fixed-size sliding window** | Compute averages over fixed range |
| **Add right, remove left** | Efficient window sum update |
