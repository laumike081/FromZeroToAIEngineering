# 1679. Max Number of K-Sum Pairs

**Difficulty**: Medium  
**Topics**: Array, Hash Table, Two Pointers, Sorting

## Problem Statement

You are given an integer array `nums` and an integer `k`.

In one operation, you can pick two numbers from the array whose sum equals `k` and remove them from the array.

Return the maximum number of operations you can perform on the array.

### Examples

**Example 1:**
```
Input: nums = [1,2,3,4], k = 5
Output: 2
Explanation: (1,4) and (2,3)
```

**Example 2:**
```
Input: nums = [3,1,3,4,3], k = 6
Output: 1
```

### Constraints

- `1 <= nums.length <= 10^5`
- `1 <= nums[i] <= 10^9`
- `1 <= k <= 10^9`

---

## Approach 1: Hash Map

### Complexity
- **Time**: O(n)
- **Space**: O(n)

````python
from collections import Counter

def max_operations(nums: list[int], k: int) -> int:
    """Hash map: O(n)"""
    count = Counter(nums)
    operations = 0
    
    for num in count:
        complement = k - num
        if complement in count:
            if num == complement:
                operations += count[num] // 2
            else:
                operations += min(count[num], count[complement])
            count[num] = 0
            count[complement] = 0
    
    return operations

# Test
print(max_operations([1, 2, 3, 4], 5))     # 2
print(max_operations([3, 1, 3, 4, 3], 6))  # 1
````

---

## Approach 2: Two Pointers (Sort First)

### Complexity
- **Time**: O(n log n)
- **Space**: O(1)

````python
def max_operations_two_pointers(nums: list[int], k: int) -> int:
    """Two pointers after sorting"""
    nums.sort()
    left, right = 0, len(nums) - 1
    operations = 0
    
    while left < right:
        total = nums[left] + nums[right]
        if total == k:
            operations += 1
            left += 1
            right -= 1
        elif total < k:
            left += 1
        else:
            right -= 1
    
    return operations

print(max_operations_two_pointers([1, 2, 3, 4], 5))  # 2
````

---

## Key Takeaways

| Pattern | When to Use |
|---------|-------------|
| **Hash map for complement** | O(n) time, O(n) space |
| **Sort + two pointers** | O(n log n) time, O(1) space |
