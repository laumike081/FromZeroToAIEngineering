## Problem Statement

Given two integers `num1` and `num2`, return the **sum** of the two integers.

### Examples

**Example 1:**
```
Input: num1 = 12, num2 = 5
Output: 17
```

**Example 2:**
```
Input: num1 = -10, num2 = 4
Output: -6
```

### Constraints

- `-100 <= num1, num2 <= 100`

---

## Solution

This is the simplest possible LeetCode problem!

````python
def sum(num1: int, num2: int) -> int:
    """O(1) time and space"""
    return num1 + num2

# Test
print(sum(12, 5))   # 17
print(sum(-10, 4))  # -6
````

---

## Bonus: Bit Manipulation (No + operator)

For interview practice, add without using `+`:

````python
def sum_bits(num1: int, num2: int) -> int:
    """Add using bit manipulation (for 32-bit integers)"""
    MASK = 0xFFFFFFFF  # 32-bit mask
    MAX_INT = 0x7FFFFFFF
    
    while num2 != 0:
        # XOR gives sum without carry
        # AND gives carry bits
        carry = (num1 & num2) & MASK
        num1 = (num1 ^ num2) & MASK
        num2 = (carry << 1) & MASK
    
    # Handle negative numbers
    return num1 if num1 <= MAX_INT else ~(num1 ^ MASK)

print(sum_bits(12, 5))   # 17
print(sum_bits(-10, 4))  # -6
````
